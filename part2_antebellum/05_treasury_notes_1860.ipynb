## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The act of December 17, 1860 (12 Statutes, 121), authorized the issue of treasury notes to the amount of \$10,000,000, payable one year after date, with interest at a rate not exceeding 12 per cent. per annum.

— Bayley, ch. 3, on the Treasury Notes of 1860.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Treasury Notes of 1860** are, in many ways, the **most extraordinary federal bond in American history** — and unmistakably the saddest. Authorized by the **Act of December 17, 1860** — three days before South Carolina seceded — they carried a punitive **9.0% coupon**, the highest interest rate ever paid on a U.S. federal bond up to that date and arguably the most expensive sovereign borrowing in American history outside of speculative episodes.

The 9% coupon tells the entire story of the **secession crisis** in a single number. By December 1860:
- South Carolina was openly preparing to secede.
- Six other Southern states were considering similar action.
- Treasury Secretary **Howell Cobb** had just resigned (December 8, 1860) to return to Georgia and support secession.
- The Buchanan administration was paralyzed and politically illegitimate.
- Federal customs receipts from Southern ports were collapsing.
- Federal credit was in tatters.

In this environment, the Treasury approached the New York banks for emergency financing. The banks refused at conventional rates. After tense negotiations, they agreed to subscribe — but only at a punitive 9% rate, well above the 6% that had been the federal norm for half a century. Even at 9%, only **\$9.95 million** was raised against the **\$10 million authorization** — a 99.5% take-up that represents banks accepting on commercial terms what they would have refused on patriotic grounds.

The Hall-Payne-Sargent database records **only 30 quantity observations** (December 1860 – March 1868) and **no secondary-market price series**. Public-holdings observations are even sparser (7 observations, December 1860 – June 1862). The Notes were retired largely on schedule by mid-1862, but the **9% rate remained a wound on federal credit** that took years to heal.


## The Secession Crisis Background

The trajectory from Lincoln's election (November 6, 1860) to Fort Sumter (April 12, 1861) is one of the most rapidly-deteriorating fiscal-political situations in American history. The Treasury Notes of 1860 sit at the center of this collapse.

### November 1860

Lincoln won the presidency on November 6 with 39.8% of the popular vote, finishing first in a four-way race. The result was understood in the South as the political triumph of a sectional party hostile to slavery. South Carolina's legislature began secession preparations immediately.

### Cobb's Resignation

Treasury Secretary **Howell Cobb** of Georgia resigned on December 8, 1860, declaring his loyalty to the South. His successor, **Philip Francis Thomas** of Maryland, served barely a month before being replaced by **John Adams Dix** of New York on January 11, 1861. The rapid turnover of Treasury Secretaries during this critical period severely damaged federal credit.

### The Failed Loan of 1860

In addition to the Treasury Notes (this bond, ID 20095), the Buchanan administration had authorized the **Loan of 1860** (Bond ID 20094) under the Act of June 22, 1860. That earlier 5%, 10-year bond had been authorized for \$21 million but ultimately placed only \$7.022 million — a placement so weak that the Treasury had to look elsewhere for emergency funds.

### The Act of December 17, 1860

The Act of December 17, 1860 authorized \$10 million in **one-year Treasury Notes** at up to 12% interest (an unprecedented ceiling). The Treasury, after negotiation, settled on 9% — high enough to motivate bank participation, but somewhat below the legal maximum.

### Why 9%?

The 9% rate reflected several factors:
- **Default risk:** Real risk that secession could trigger federal default.
- **Liquidity premium:** A small market for emergency federal paper.
- **Market power of New York banks:** With Southern banks effectively unavailable, NY banks had monopoly pricing power.
- **Political risk premium:** Banks knew they were lending into a deeply uncertain political environment.

### The Banking-Suspension Crisis

In December 1861 — exactly twelve months after this Note's authorization — Northern banks would suspend specie payments for the first time since 1838. The 9% Notes are best understood as a precursor to that broader liquidity crisis: an early signal that the antebellum financial system could not survive the political shock of secession.

## Quantity Outstanding

The Notes peaked at \$9.95 million in March 1861 — within months of authorization. Although the database first-issue date is recorded as June 30, 1861, substantial issuance had already occurred during December 1860 through March 1861 (the Hall-Payne-Sargent database appears to use an aggregated first-issue marker for the bond's "active" period rather than the actual first specific issuance date).

The Notes were retired through 1862 as their one-year maturities came due. Public-holdings data ends in mid-1862, indicating that by then most of the Notes had been redeemed by their bank holders. The long total-outstanding tail through 1868 reflects small residual amounts of unredeemed or replaced certificates.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1860-12-20'), pd.Timestamp('1861-04-12'), alpha=0.15, color='#f39c12', label='Secession Crisis')
ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20095, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Treasury Notes of 1860 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20095, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.2f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.3f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Price Data

The Hall-Payne-Sargent database records no secondary-market price series for these Notes. As short-term at-maturity instruments, they typically did not develop deep secondary markets — bank holders bought them, held them, and redeemed them at maturity. The absence of price data is therefore consistent with the instrument's character.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20095, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1860-12-20'), pd.Timestamp('1861-04-12'), alpha=0.15, color='#f39c12', label='Secession Crisis')
    ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Treasury Notes of 1860 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Treasury Notes of 1860 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Legacy and Significance

### A 9% Monument to Crisis

The Treasury Notes of 1860's 9% coupon stands as a numerical monument to the **fiscal collapse of antebellum federal authority**. No federal bond before 1860 had paid 9% in peacetime; no federal bond after 1860 (with the partial exception of some Compound Interest Notes in 1864) would either. The 9% rate is a unique data point — a specific, quantifiable measure of how completely the secession crisis had impaired federal sovereign credit by December 1860.

For comparison:
- The **Six Per Cent Stock of 1790** (Bond ID 20021) carried 6% — the original Hamilton standard.
- The **Loan of 1846** (Bond ID 20086) carried 6% during the Mexican War.
- The **Loan of 1848** (Bond ID 20090) carried 6% with a 3% issuance premium.
- The **Six Per Cent Loan of 1812** (Bond ID 20048) carried 6% but issued at par in patriotic placement.
- The Civil War Five-Twenties of 1862 (Bond ID 20101) carried 6%.

The Treasury Notes of 1860 break this 6% pattern definitively — and reflect the fact that, by late 1860, **the conventional Hamiltonian framework for U.S. sovereign borrowing had temporarily ceased to function**.

### A Catalyst for Reform

The trauma of the 9% Notes was a major factor in the rapid construction of new financial infrastructure during the Civil War:

- **Legal Tender Act** (February 1862): Created greenbacks, providing the Treasury a way to expand the money supply directly.
- **Five-Twenties of 1862** (Bond ID 20101): Set up Jay Cooke's mass-marketing campaign that would democratize federal bond ownership.
- **National Banking Act** (February 1863): Created the National Banking System, with mandatory federal-bond reserve requirements.

Each of these reforms aimed, in part, at preventing future episodes where the federal government would be at the mercy of New York banks for emergency financing. The 9% Notes of 1860 were Exhibit A in the case for institutional reform.

### A Quiet Antebellum Coda

Symbolically, the Treasury Notes of 1860 are the **last antebellum federal bond**. Within months of their issuance, the entire framework of antebellum federal finance — based on coin payments, par-value bond placements, modest borrowing, peacetime balanced budgets — would be swept away by Civil War necessity. Greenbacks, mass-marketed bonds, paper money, and a National Banking System would replace it. The Treasury Notes of 1860 stand at the dividing line between two completely different American fiscal eras.
