## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The act of September 9, 1850 (9 Statutes, 446), authorized the issue of \$10,000,000 of stock, bearing interest at the rate of 5 per cent. per annum, in part payment of the claims of Texas for the surrender of her claim to a large portion of the territory of New Mexico.

— Bayley, ch. 3, on the Texas Indemnity Stock.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Texas Indemnity Stock** is one of the most politically consequential federal bonds in American history. Authorized by the **Act of September 9, 1850** as part of the **Compromise of 1850** — the famous package of five bills designed to defuse the brewing sectional crisis over slavery in the western territories — it served the very specific purpose of **buying out Texas's pre-annexation debt and territorial claims** in exchange for Texas's acceptance of a smaller, defined western boundary.

The story is intricate. When Texas joined the Union in 1845, it brought with it approximately **\$10 million in debt** accumulated during its decade as the independent Republic of Texas (1836–1845). Texas also claimed extensive territory in what is now eastern New Mexico — claims that had no clear legal foundation but that Texas was prepared to enforce militarily. By 1850, Texas was on the brink of armed conflict with federal authorities in New Mexico, with serious risk of triggering wider sectional war.

The compromise solution: **the United States would pay Texas \$10 million in bonds in exchange for Texas dropping its New Mexico claims**. Texas would use the bond proceeds to retire its old debt; New Mexico would become a U.S. territory; and the federal government would absorb the political cost of the bond issue. Both Daniel Webster (Massachusetts Whig) and Henry Clay (Kentucky Whig, the principal architect of the Compromise) saw the Texas Indemnity as the **financial linchpin** that made the entire 1850 Compromise possible.

In the Hall-Payne-Sargent database, the Texas Indemnity Stock has **120 monthly price observations** (April 1852 – December 1864). Mean price: **\$106.29**. Range: **\$85.00** (January 1862, deep Civil War uncertainty) to a startling **\$200.00** (November 1864, just before final redemption — apparently a thinly-traded extreme price). Peak outstanding: **\$5.00 million** (March 1851).


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> It provided that if the state of Texas should consent that her boundary on the north should commence at a point at which the meridian of one hundred degrees west from Greenwich is intersected by the parallel of 36 degrees 30 minutes north latitude and run thence due west to the meridian of one hundred and three degrees west from Greenwich, then south to the thirty-second degree of north latitude, then on said parallel to the Rio Bravo del Norte, and thence, with the channel of said river, to the Gulf of Mexico, and should cede to the United States all her claim to the territory exterior to these boundaries, and relinquish all claim on the United States for liability for her debts, and for compensation for the surrender of her ships, forts, custom-houses, public buildings, &c., then the United States, in consideration for the same, should pay her the sum of \$10,000,000 in stock, bearing 5 per cent. interest, payable semi-annually, redeemable at the end of fourteen years.

— *Bayley, ch. 3 (Texas Indemnity Stock), p. 164 — Act of September 9, 1850 (9 Statutes, 447).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> It was said, in debate on the subject, that her southeast corner was at the mouth of the Rio Grande, the region of perpetual flowers, her northeast corner near the South pass in the Rocky mountains, a region of everlasting snows; that she had a front on the Gulf of Mexico of a thousand miles, a front on the Rio Grande, partly disputed, of 2,000 miles, an undisputed frontier on the Arkansas of 1,000 miles, and within these boundaries she claimed an area of 350,000 square miles, or more than equal to seven states as large as Pennsylvania.

— *Bayley, ch. 3 — Senate debate of August 1850 describing the scope of Texas's pre-1850 territorial claim.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Under this act but \$5,000,000 in stock was issued, as a subsequent act, approved February 28, 1855 (10 Statutes, 617), provided that the creditors of Texas should be paid in cash, and increased the amount to be paid them to \$7,750,000, to be divided pro rata among such creditors of the State as were comprehended in the act of September 9, 1850.

— *Bayley, ch. 3 — only \$5M of the \$10M authorization was issued in stock; the 1855 act converted the second tranche into a cash settlement to Texas creditors.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Compromise of 1850

The Compromise of 1850 was a five-part legislative package brokered by Henry Clay (and, after Clay's withdrawal due to illness, by Stephen Douglas) to address the sectional tensions arising from the Mexican Cession of 1848. The five bills:

1. **California admitted as a free state** (September 9, 1850)
2. **New Mexico organized as a territory** with no slavery prohibition (popular sovereignty)
3. **Utah organized as a territory** with no slavery prohibition (popular sovereignty)
4. **Fugitive Slave Act** strengthened (forced free states to assist in returning escaped slaves)
5. **Texas Boundary settlement and Indemnity** (this bond's authorizing act)

The Texas Indemnity was, in some ways, the most concrete element of the package. While the other four bills involved abstract questions of state-formation and federal power, the Texas Indemnity was a specific cash transaction with measurable terms.

### Why \$10 Million?

The number was negotiated. Texas's pre-annexation debt totaled approximately **\$10 million** (including interest accrued since the 1845 annexation). Texas creditors — many of them New York and Philadelphia merchants — had been pressing for federal assumption of these debts for years, and the Compromise gave them their chance. The \$10 million figure was set both to fully retire the pre-annexation debt and to provide enough cushion that Texas would accept the loss of its New Mexico claims.

### The Bond Mechanism

Rather than pay Texas in cash, the federal government issued **Texas Indemnity Stock** — 5%, 14-year, semi-annual, coin-payable bonds — directly to Texas. Texas then used these bonds to pay off its creditors, who held them as standard federal obligations.

This had several advantages:
- **No immediate cash strain on the Treasury:** The federal government did not have to find \$10 million in cash; it simply issued bonds.
- **Marketability:** The bonds could be traded by Texas creditors on standard secondary markets.
- **Defined terms:** The 14-year, 5% structure was straightforward and easily understood.

## Quantity Outstanding

The Texas Indemnity Stock peaked at \$5 million in March 1851, slightly below the full \$10 million authorization (the remainder was held back due to a holdback clause requiring Texas creditors to release certain claims before the second tranche could be issued; some of these claims were never resolved). After issuance, the Stock held essentially flat through the 1850s. Beginning around 1860, the Treasury used Sinking Fund authority to retire portions of the Stock, accelerating into the early 1860s. By the redemption date (January 1865), most of the Stock had been retired.

The very long retirement tail (database observations through 1891) reflects small unredeemed balances and disputed claims that took decades to resolve.

## Secondary Market Price

The Texas Indemnity Stock's price history mostly tracks the broader federal bond market through the 1850s and early 1860s, with two distinctive features worth noting:

1. **The 1862 Trough (\$85):** Like most federal bonds, the Texas Indemnity Stock fell sharply in early 1862 as Civil War uncertainty peaked. The trough is somewhat deeper than for the contemporaneous Loan of 1848 (Bond ID 20090, low \$88), reflecting both the smaller market for this less-liquid issue and lingering political controversy over its association with the Compromise of 1850 (which Confederate-state observers viewed retrospectively as a betrayal of Southern interests).

2. **The 1864 Peak (\$200):** The November 1864 peak appears to be a thinly-traded extreme price. By late 1864, the redemption date (January 1865) was approaching, the bond was being aggressively retired, and very few were available on the market. The \$200 quote — twice par — likely reflects either a single transaction at a distorting price, or a quote influenced by greenback depreciation against gold (since the Stock was coin-payable, it commanded a substantial premium against gold-depreciated greenbacks at this stage of the war).

For a more typical price reading, the **mean of \$106.29** over 120 observations is more representative.

## Legacy and Significance

### Ten Million Dollars That Postponed a War

The Texas Indemnity Stock is, in a deep sense, **the bond that bought time**. By resolving the Texas-New Mexico boundary dispute and removing the immediate flashpoint for armed sectional conflict, it gave the United States a decade of additional peace. That decade was itself precarious — Bleeding Kansas, Dred Scott, Brown's raid all threatened to ignite war in the late 1850s — but it also provided crucial time for Northern industrial development, Western settlement, and political mobilization that ultimately favored the Union side when civil war did come.

It is hard to overstate the significance. Without the Compromise of 1850 (and without this Stock as part of it), war between Texas and federal authorities in New Mexico could plausibly have begun in late 1850 — eleven years before Fort Sumter. A war beginning in 1850, before the Republican Party existed, before Northern industrial dominance was secured, and before the railroads connected the Northwest to Eastern markets, might well have ended very differently.

### A Template for Federal-State Settlements

The Texas Indemnity established a template that would be used in later federal-state debt settlements. The most direct successor is the **Funding Act of 1870** (under which old Civil War 6% debts were refinanced into 4%, 4.5%, and 5% bonds, including the 4.5% Loan of 1891 and the 4% Loan of 1907 — Bond IDs 20120 and 20121). More distantly, the post–Spanish-American War debt-restructuring operations of the early 20th century used similar techniques.

### A Quiet Antebellum Survivor

Unlike the Loan of 1848, which has a famous price history visible across the antebellum boom, the Texas Indemnity Stock is rarely discussed in financial history. But it played its political role faithfully and was retired largely on schedule. Its primary monument is not its financial impact but its **political achievement**: ten million dollars that may have postponed civil war by a decade.
