## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of July 22, 1846 (9 Statutes, 39), the Secretary of the Treasury was authorized, with the consent of the President, to cause to be issued United States stock for the sum of \$10,000,000.

— Bayley, ch. 3, on the Loan of 1846.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Loan of 1846** is the principal funding instrument of the **Mexican-American War** (1846–1848). Authorized by the **Act of July 22, 1846** — three months after Congress declared war on Mexico — it carried a 6% semi-annual coupon, a 10-year term, and was marketed at par. The Treasury issued \$5 million under this authority, all of which was placed by the autumn of 1847.

The Loan came at a curious moment in American fiscal history. The Independent Treasury Act of 1846 — which separated federal finances from the private banking system — had taken effect in August, just weeks after the Loan's authorizing act. This meant that the Treasury was conducting **its first major war-bond placement under the new "subtreasury" system**, which required all federal receipts and payments to be in specie (gold or silver coin). For bond subscribers, this meant they had to deliver actual gold or silver coin to subtreasury depositories, rather than the bank notes that had typically funded earlier bond issues.

Despite this novel constraint, the Loan placed successfully. The peak outstanding (\$5 million) was reached by September 1847, well within the war years. Secondary-market price was strong: **101 monthly observations** in the database (December 1846 – October 1855), mean **\$105.93**, range **\$96.00–\$110.44**. The bond traded at premium throughout most of its life, supported by the modest scale of the Mexican War (a fraction of the financial burden of the War of 1812 or the eventual Civil War) and the strength of the post-1849 American economy after the California Gold Rush began.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> The second section of the act of July 22, 1846 (9 Statutes, 40), authorized the President, if, in his opinion, it should be for the interest of the United States so to do, instead of issuing the whole amount of treasury notes authorized by the first section, to borrow, on the credit of the United States, such a sum as he might deem proper, but not exceeding, together with the treasury notes issued, the sum of \$10,000,000.

— *Bayley, ch. 3 (Loan of 1846), p. 112 — Act of July 22, 1846 (9 Statutes, 40).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> War with Mexico was declared May 13, 1846. On the 15th of June the Secretary of the Treasury informed Congress that if the war should continue until July 1, 1847, there would be a deficiency in the treasury of \$12,587,000. To prevent this threatened deficiency, by the issue of treasury notes, a bill was introduced in the House of Representatives July 6, 1846, passed both houses without much debate, and was approved July 22, 1846 (9 Statutes, 39).

— *Bayley, ch. 3 — context of the war declaration (May 13, 1846) and the Treasury's deficit estimate.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> The stock was to be issued under the limitations and provisions of the act of April 15, 1842, and to be redeemable at a period not exceeding ten years from the date of issue. No commissions were to be paid to agents. Under this act the sum of \$4,999,149 45 was issued.

— *Bayley, ch. 3 — final issuance amount (\$4,999,149.45 against the \$5M cap, with the rest of the \$10M filled by Treasury Notes), and the ten-year redemption window.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## Polk and the Mexican War

President **James K. Polk** had run for office in 1844 on an aggressive expansionist platform: annexation of Texas, settlement of the Oregon dispute with Britain, and acquisition of California. The Texas annexation (December 1845) precipitated tensions with Mexico, which had never recognized Texan independence. In April 1846, Polk dispatched General Zachary Taylor to the disputed Texas-Mexico border. A Mexican cavalry skirmish with American troops on April 25, 1846 became Polk's casus belli.

### The Financing Strategy

Polk's Treasury Secretary **Robert J. Walker** (a Mississippi Democrat and architect of the lower 1846 tariff) developed a financing strategy with three legs:

1. **Tariff revenues:** The Walker Tariff of 1846, despite lowering rates, produced strong revenues thanks to a booming economy.
2. **Treasury Notes:** Short-term interest-bearing notes (Bond ID 20085, the Treasury Notes of 1846) handled day-to-day cash flow.
3. **The Loan of 1846:** Long-term capital, sold at par to absorb major capital expenditures (especially General Winfield Scott's Vera Cruz–Mexico City campaign).

### A Smaller War — A Smaller Loan

The Mexican War cost the United States approximately **\$75 million** in direct expenditures — large by 1840s standards but small by Civil War standards (which would eventually cost ~\$3 billion). The Loan of 1846 raised only \$5 million; the rest came from tariff revenues, Treasury Notes, and a follow-on Loan of 1847 (Bond ID 20089, also \$18.4 million). The relatively small size of the war loan reflected the manageable scale of the conflict and the strength of antebellum federal finances.

## Quantity Outstanding

The Loan reached its \$5 million peak by September 1847 — within fifteen months of authorization. After the war ended in February 1848, the Treasury began retiring the Loan systematically. By 1856 (the redemption window), most of the issue had been retired, with the remaining small balance settled in the years just before the Civil War.

## Secondary Market Price

The Loan of 1846's price history is one of relatively quiet stability — a useful reminder that not every U.S. bond era was dominated by crisis:

1. **December 1846 – January 1848 (~\$96-104):** Wartime fluctuations in a narrow range. The brief sub-par dip in early 1848 reflects skepticism about the final cost of the war and the terms of peace.

2. **February 1848 onward:** The Treaty of Guadalupe Hidalgo and the discovery of gold in California (within days of each other!) transformed the financial outlook. Gold was about to start flowing into the federal treasury and the broader economy.

3. **1849–1852 Premium Build:** Gold Rush wealth, robust customs revenues, and confidence in long-term American growth pushed prices steadily upward.

4. **November 1852 Peak (\$110.44):** The lifetime high. Prices reflected a combination of strong federal finances, declining market interest rates, and a generally optimistic mid-19th-century investment environment.

5. **1853–1855 Modest Decline:** As the redemption date approached, prices began converging back toward par.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20086, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1846-04-25'), pd.Timestamp('1848-02-02'), alpha=0.15, color='#e74c3c', label='Mexican-American War')
    ax.axvspan(pd.Timestamp('1857-08-24'), pd.Timestamp('1859-12-31'), alpha=0.15, color='#9b59b6', label='Panic of 1857')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Loan of 1846 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Loan of 1846 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20086, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### Successful War Finance Without Major Stress

The Loan of 1846 demonstrates a model of **measured wartime sovereign borrowing**. Unlike the troubled placement of the Six Per Cent Loan of 1812 (Bond ID 20048) or the heroic last-minute rescue required for the Sixteen Million Loan of 1813 (Bond ID 20052), the Mexican War loan placed smoothly, traded at premium, and was retired on schedule. This relative ease reflects the enormous progress in American financial institutions and credibility between 1813 and 1846.

### A Test of the Independent Treasury

The Loan was the first major federal bond placement under the **Independent Treasury system** that took effect in August 1846. By requiring specie payments, the system imposed real costs on bond subscribers, but ultimately did not prevent successful placement. This was a vindication of the Polk-Walker design and confirmed that federal finance could function without a central bank as long as institutional rules were clear and consistent.

### Acquisition Bonds

The Mexican War, paid for partly through this Loan, resulted in the **Mexican Cession** — California, New Mexico, Arizona, Utah, Nevada, and parts of Colorado and Wyoming. Combined with the Louisiana Purchase (financed by Bond ID 20044) and the Texas annexation, the Mexican Cession completed America's continental footprint. The Loan of 1846, while modest in itself, is therefore part of the larger story of how **federal bonds funded the territorial expansion of the United States**.

### A Contrast with the Civil War to Come

The smooth, almost routine placement of the Loan of 1846 stands in stark contrast to the desperate financial improvisation that would characterize the early Civil War. The fifteen years between the Mexican War and Fort Sumter would see the slow erosion of federal financial capacity — first the Panic of 1857, then the secession crisis of 1860–1861 — until by April 1861 the new Lincoln administration would inherit a Treasury in shambles. The Loan of 1846 represents the **last successful war finance under the antebellum framework**.
