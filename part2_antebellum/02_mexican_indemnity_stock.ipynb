## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The act of August 18, 1846 (9 Statutes, 96), authorized the issue of stock to the amount of \$303,573, in payment of the indemnity due to citizens of the United States from Mexico.

— Bayley, ch. 3, on the Mexican Indemnity Stock.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Mexican Indemnity Stock** is a small but historically interesting bond that illustrates a distinctive 19th-century practice: the use of federal sovereign debt to compensate American citizens for **losses suffered abroad** when the responsible foreign government was unable or unwilling to pay directly.

The story begins in 1839, before the Mexican-American War. Mexico had agreed (under the Convention of April 11, 1839) to pay approximately \$2 million to American citizens whose property had been seized or damaged by Mexican authorities during the political upheavals of the 1820s and 1830s. Mexico paid a few installments, then defaulted in 1842. The U.S. government, having endorsed the underlying claims through the Convention, faced a dilemma: it could either pursue Mexico through diplomatic and (potentially) military means, or it could **assume responsibility for the unpaid balances itself** and issue federal bonds to settle them.

The **Act of August 18, 1846** — passed during the early months of the Mexican-American War — chose the second approach. It authorized the Treasury to issue **Mexican Indemnity Stock** to American claimants, with a 5% coupon and 5-year term. The Stock served as a **substitute payment**: claimants received federal bonds rather than cash, and the bonds could be sold on the secondary market for whatever the bondholder could obtain.

The peak outstanding was small — only **\$0.30 million** — reflecting the limited universe of qualifying claimants. Quantity data shows 292 monthly observations through 1886, indicating extremely slow retirement (some claims took decades to resolve). The Hall-Payne-Sargent database does not record secondary-market price data for this Stock.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> By the terms of the convention of January 30, 1843, between the United States and the Mexican republic, it was provided that on the 30th day of April, 1843, the Mexican government should pay all the interest which might then be due on the awards in favor of claimants under the convention of the 11th of April, 1839. The principal of the said awards, and the interest accruing thereon, to be paid in five years, in equal installments every three months, to commence April 30, 1843. These payments were to be made in the city of Mexico, to such person as the United States might authorize to receive them, in gold or silver money.

— *Bayley, ch. 3 (Mexican Indemnity Stock), p. 117 — terms of the 1843 convention with Mexico.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> It appears from the correspondence relative to this affair that, for the avowed purpose of liquidating these claims, Santa Ana, the head of the Mexican government, in May, 1843, decreed the collection of a forced loan, to be distributed in certain proportions through the various departments of the Mexican republic, and paid at periods corresponding to those stipulated by the convention. This measure, essentially unpopular, was imperfectly enforced, while the temptation to a misapplication of the funds collected, amidst the difficulties by which Santa Ana was surrounded, was very great. Under these circumstances the Mexican government was unable to pay the installment which became due in April, 1844; and in July of the same year, when another installment became due, the incapacity of the government had become still clearer.

— *Bayley, ch. 3 — the Santa Ana "forced loan" episode that led Mexico to default on the indemnity payments.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> In 1846, the United States being at war with Mexico, a clause was inserted in the first section of the civil and diplomatic appropriation act of August 10 (9 Statutes, 94), providing: "For paying the principal and interest of the fourth and fifth installments of the Mexican indemnities due in April and July, in the year 1844, the sum of \$320,000: Provided, The claimants, each for himself, shall relinquish to the United States his right to said installments: Provided further, That each of the claimants shall agree to take in payment the scrip of a stock bearing interest at five per cent., payable in five years." Under the authority thus granted, five per cent. stock, amounting to \$303,573 92, was issued.

— *Bayley, ch. 3 — the precise authorization (Aug 10, 1846) and final issuance figure (\$303,573.92) of the Mexican Indemnity Stock.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Convention of 1839 and Its Aftermath

The American claims that the Mexican Indemnity Stock was designed to settle had a complicated history. During Mexico's tumultuous early decades — multiple revolutions, the Texan revolt, repeated changes of government — Americans operating in Mexico frequently lost property, ships, or cargoes to revolutionary armies, customs disputes, or arbitrary expropriations. By the late 1830s, U.S. citizens had filed claims totaling over \$6 million against the Mexican government.

### The Mixed Claims Commission

Following diplomatic pressure (and President Andrew Jackson's threat of force), Mexico agreed in 1838 to submit the claims to a **Mixed Claims Commission** — a binational tribunal that would adjudicate which claims were valid. The Commission, after lengthy proceedings, awarded American claimants approximately **\$2 million**.

### The Convention of April 11, 1839

Under the resulting Convention, Mexico agreed to pay the \$2 million in monthly installments. Several payments were made between 1839 and 1842, but Mexico — perpetually short of cash and beset by political turmoil — defaulted on the remaining balance in 1842. By 1846, the unpaid balance was approximately **\$3.25 million** (including accrued interest).

### Why a U.S. Bond?

The decision to issue federal bonds rather than press Mexico further had several rationales:

1. **Practical resolution:** American claimants had been waiting nearly a decade and were unlikely to receive payment from Mexico under any circumstances.
2. **Political symbolism:** As the war with Mexico began, settling old claims domestically (rather than continuing to negotiate with a hostile power) made political sense.
3. **Treaty leverage:** The U.S. could now treat the claims as having been satisfied by the federal government, thereby strengthening its eventual peace-treaty negotiating position (the Treaty of Guadalupe Hidalgo would explicitly transfer responsibility for these claims to the U.S. government in exchange for Mexican territorial concessions).

The Mexican Indemnity Stock thus functioned as both a **debt-settlement instrument** and a **diplomatic pre-positioning move**.

## Quantity Outstanding

The Mexican Indemnity Stock peaked at \$300,000 in June 1848 — a tiny issue by any standard. The very long retirement tail (extending into the late 1880s) reflects how slowly some of the underlying claims were ultimately resolved: many small claimants either lost track of their bonds, died without proper estate handling, or had legal disputes about title that took decades to resolve. The Stock thus had an unusually long quiet life on Treasury books.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1846-04-25'), pd.Timestamp('1848-02-02'), alpha=0.15, color='#e74c3c', label='Mexican-American War')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20087, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Mexican Indemnity Stock — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20087, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.2f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.3f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Legacy and Significance

### A Distinctive Instrument Class

The Mexican Indemnity Stock is one of relatively few American sovereign instruments specifically designed to compensate citizens for foreign losses. Other examples in the broader Hall-Payne-Sargent database include the **Texas Indemnity Stock** (Bond ID 20093, see following chapter), various claims-settlement bonds issued after the Spanish-American War, and certain WWI-era reparations-related instruments. As a class, these bonds reveal an important dimension of 19th-century American foreign relations: **the federal government routinely backstopped private claims against foreign sovereigns, using its own credit to make whole what foreign governments would not pay**.

### A Fiscal Footnote to a Major War

The Stock's modest size — under one-third of a million dollars at peak — makes it easy to overlook. But its significance is symbolic rather than financial. It marks the federal government's assumption of responsibility for **citizen claims abroad**, a practice that would grow substantially in later decades as American commercial interests expanded globally. Every modern claims-resolution program — from the Iran Hostage Compensation to post-9/11 settlements — has institutional ancestry in instruments like this 1847 Mexican Indemnity Stock.

### Closure Through the Treaty of Guadalupe Hidalgo

The Treaty of Guadalupe Hidalgo (February 2, 1848) explicitly transferred responsibility for the underlying claims from Mexico to the U.S. government, in effect retroactively ratifying the 1846 Act's approach. In exchange, Mexico ceded to the United States approximately 525,000 square miles of territory (modern California, Nevada, Utah, Arizona, New Mexico, and parts of Colorado and Wyoming) and accepted the Rio Grande as the Texas border. The Mexican Indemnity Stock thus represents one small but tangible piece of the financial machinery that made the **Mexican Cession** possible.
