## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of March 31, 1848 (9 Statutes, 217), the Secretary of the Treasury was authorized, with the consent of the President, to borrow on the credit of the United States such sum or sums, not exceeding sixteen millions of dollars, as he might find necessary.

— Bayley, ch. 3, on the Loan of 1848.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Loan of 1848** is the largest antebellum federal long-bond and one of the most significant pre–Civil War sovereign debt instruments. Authorized by the **Act of March 31, 1848** — almost exactly two months after the Treaty of Guadalupe Hidalgo ended the Mexican-American War — it raised funds both to **complete payment of the Mexican Cession** (\$15 million payable to Mexico under the Treaty's terms) and to refund earlier short-term war debt at more favorable long-term rates.

The Loan was issued at a slight premium (\$103) — a striking contrast to the discounted issuances of the War of 1812. By 1848, American sovereign credit had recovered from the Polk-era fiscal stress, the war was over, gold was about to start flowing from California, and investors were eager for high-quality long-dated paper. The premium reflects this favorable environment.

The Loan carried a **20-year term** (long by 1840s standards), a 6% coupon, and was redeemable after 1868. Peak outstanding reached **\$15.74 million** (September 1849) — about as much as the entire Mexican War had cost the federal government in cash terms.

The Hall-Payne-Sargent database records **168 monthly price observations** (November 1848 – December 1865). Mean price: **\$113.46**. Range: **\$88.00** (August 1861, the early Civil War nadir) to **\$133.56** (December 1859, the antebellum boom peak). The Loan thus has the unusual distinction of being a single bond whose price history **spans the entire antebellum boom, the secession crisis, and the early Civil War** — a remarkable lens through which to view the era.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> The bill passed both houses after an extended debate and was approved March 31, 1848 (9 Statutes, 217). It authorized the President to borrow, on the credit of the United States, at any time within one year from the passage of the act, a sum not exceeding \$16,000,000, at an interest of not more than 6 per cent. per annum, payable quarterly or semi-annually, reimbursable at any time after July 1, 1868. No certificate of stock was to be issued for a less amount than \$50.

— *Bayley, ch. 3 (Loan of 1848), p. 153 — Act of March 31, 1848 (9 Statutes, 217).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> The President of the United States, in his message of December 7, 1847, informed Congress that if the war with Mexico continued until July 1, 1848, the expenditure would probably exceed the revenue provided by law about \$16,000,000, while the Secretary of the Treasury, in his report of December 8, estimated that, adding the sum necessary to be kept in the treasury to meet the wants of the government and maintain the enlarged operations at the mints, there would be required \$18,729,114 27, in addition to the probable revenue. A bill to authorize a loan not to exceed \$18,500,000 was introduced in the House of Representatives January 19, 1848.

— *Bayley, ch. 3 — Polk's December 1847 message and the Treasury's deficit estimate that produced the loan authorization.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> The chairman of the committee of ways and means informed the House, during the debate, that an error had been discovered in the financial statement of the Secretary of the Treasury, which disclosed the gratifying fact that there was in the treasury nearly \$7,000,000 more than had been stated to the House, in consequence of which it was supposed the loan might be reduced to \$12,000,000. … Under this act stock to the amount of \$16,000,000 was issued and sold at a small premium.

— *Bayley, ch. 3 — the legislative debate (including the discovered \$7M accounting error) and the final premium-priced placement of the full \$16M.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## A Bond for a New Era

The Loan of 1848 is best understood as a **transition instrument** between two distinct phases of American fiscal life. It was authorized at the close of the Mexican War and lived through:

- **The Mexican Cession completion** (1848–1850): Final payments to Mexico under the Treaty of Guadalupe Hidalgo.
- **The California Gold Rush** (1849–1855): Massive specie inflows transformed the American monetary base.
- **The Compromise of 1850**: Including the **Texas Indemnity** (Bond ID 20093, see following chapter), which absorbed Texas's pre-annexation debt and helped defuse sectional tensions.
- **The 1850s Antebellum Boom**: Expanding railroads, surging cotton exports, immigration-fueled population growth.
- **The Panic of 1857**: A sharp but relatively brief financial crisis triggered by the failure of the Ohio Life Insurance and Trust Company.
- **The Sectional Crisis** (1858–1860): Bleeding Kansas, Dred Scott decision, Lincoln-Douglas debates, John Brown's raid.
- **The Secession Crisis** (December 1860 – April 1861): Seven Southern states leave the Union; Buchanan administration nearly bankrupt.
- **The Early Civil War** (April 1861 – December 1865): Initial military reverses; emergency federal financing; eventual Union victory.

That a single 1848 bond's price history can speak to all of this is testament to both its long maturity and the unusual continuity of secondary-market data for this issue.

## Secondary Market Price

The Loan of 1848's price history is one of the most informative in the Hall-Payne-Sargent database for understanding antebellum financial markets:

1. **November 1848 – 1851 (\$100-110):** Quiet stable trading near the issuance premium. The combination of post-war fiscal strength and California gold inflows kept prices steady.

2. **1852–1856 (\$110-120):** Antebellum boom. Robust customs revenues, expanding economy, and (after 1856) the maturity of much earlier antebellum debt drove prices into a sustained premium.

3. **August 1857 Panic Drop:** The Panic of 1857 produced a sharp but brief decline. The crisis was triggered by the failure of the Ohio Life Insurance and Trust Company (August 24, 1857) and rapidly affected New York and Philadelphia banks. The Loan dipped, but not catastrophically — federal credit remained sound even when private credit was in crisis.

4. **1858–1859 Recovery and Peak (\$133.56 in December 1859):** As the Panic eased and the economy recovered, prices climbed to the **lifetime peak**. The peak coincides eerily with the deepest stage of the slavery-secession debate: John Brown was hanged on December 2, 1859; the Senate was deadlocked over Speakership for weeks; the Charleston Democratic Convention's collapse was just months away. Yet bond markets, oddly, remained sanguine.

5. **1860 Decline:** Prices began drifting lower as Lincoln's election became likely. Investors rationally anticipated that a Lincoln victory would precipitate Southern secession.

6. **December 1860 – April 1861 Secession Crisis:** Sharp price declines as Southern states left the Union, federal authority collapsed in the South, and Buchanan administration paralysis became total. The Loan traded below par (~\$95) for the first time since 1849.

7. **August 1861 Lifetime Low (\$88):** After Fort Sumter and Bull Run, with the war's outcome genuinely uncertain. This was the lowest price the Loan ever traded.

8. **1862–1865 War Recovery:** Once Union financing was placed on a firmer footing (Legal Tender Act, Five-Twenties of 1862, National Banking Act), prices recovered. By 1865, the Loan was again trading in the \$105-115 range.

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20090, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### A Premium-Priced War Bond

The Loan of 1848's \$103 premium issuance is a striking statistic. It tells us that, by 1848, **American sovereign credit had recovered to such an extent that the Treasury could place war-period debt at premium rather than discount**. This is a dramatic improvement from the \$88 discount of the Sixteen Million Loan of 1813 (Bond ID 20052), and reflects four decades of careful fiscal management since Hamilton's day.

### A Lens on the Antebellum Boom

The Loan's secondary-market price history provides one of the cleanest windows into the **financial dynamics of antebellum America**. The slow climb from \$100 (1848) to \$133 (1859) traces the underlying economic boom of the 1850s — railroad construction, gold-fueled monetary expansion, agricultural specialization, immigration, urbanization. The brief Panic of 1857 dip is the only serious interruption.

### A Premonition of Civil War

The price decline through 1860 and into 1861 is one of the rare examples in early-American financial history where **bond markets clearly anticipated political catastrophe**. The decline began before South Carolina's secession, accelerated through the Buchanan-Lincoln transition, and bottomed during the war's earliest months. This is no different in spirit from modern sovereign-bond markets pricing in war risk — but in 1860 it was a relatively new financial phenomenon, and the data show that markets did anticipate disaster.

### A Survivor

The Loan of 1848 ultimately did its job. It financed the Mexican Cession completion, weathered the antebellum cycles, and was retired on schedule under the Funding Act-era refundings of the late 1860s. By the time it was fully redeemed in the early 1870s, it had been on Treasury books for nearly a quarter-century — an unusually long life for a single antebellum bond, and one that gave researchers an extraordinarily rich price record.
