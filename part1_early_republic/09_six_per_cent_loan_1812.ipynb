## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> On the 14th of March, 1812 (2 Statutes, 694), an act was approved authorizing the President to borrow eleven millions of dollars at a rate of interest not exceeding 6 per cent.

— Bayley, ch. 2, on the Six Per Cent Loan of 1812.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Six Per Cent Loan of 1812** is the first major federal bond issued for the **War of 1812**. Authorized by the **Act of March 14, 1812** — three months *before* the actual declaration of war — it represented the Madison administration's belated recognition that the looming conflict with Britain would require substantial new borrowing. With its **12-year term, 6% quarterly coupon**, and **par-value placement**, it adhered to the Hamilton template that had defined American sovereign debt for two decades.

But the financial environment in 1812 was vastly different from the optimistic moment of 1804 (Louisiana Purchase) or 1807 (Gallatin's refunding operations). The **First Bank of the United States had been allowed to expire in 1811**, depriving the Treasury of its primary financial agent. The **Embargo of 1807–1809** had crushed customs revenues. The British Royal Navy was at the peak of its power. And the underlying political support for the war was deeply divided — most New England Federalists openly opposed it, and many refused to subscribe to war bonds.

The result was a difficult issuance experience. Of the **\$11 million authorized**, only **\$7.81 million** was ultimately taken up — a 71% take-up rate that was disappointing by Gallatin's standards. The bonds were sold at par, but only by accepting subscriptions in depreciated state bank notes (which traded below specie).

In the Hall-Payne-Sargent database, this Loan has **150 monthly price observations** (April 1813 – September 1825). Mean price: **\$99.96**. Range: **\$75.00** (January 1815, the war's darkest hour) to **\$109.50** (June 1821, the post-war recovery peak). Peak outstanding: **\$7.81 million** (June 1815).


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> It authorized the President to borrow, on the credit of the United States, \$11,000,000, at a rate of interest not exceeding 6 per cent., payable quarter-yearly, the money to be reimbursable, at the pleasure of the United States, at any time after the expiration of twelve years from January 1, 1813. By a supplemental act, approved July 6, 1812 (2 Statutes, 784), the Secretary of the Treasury was authorized to employ an agent or agents, for the purpose of selling any part of this stock.

— *Bayley, ch. 2 (Six Per Cent. Loan of 1812) — Act of March 14, 1812 (2 Statutes, 694).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> The bills which had passed or were likely to pass Congress to put the country in readiness for war called for an extraordinary expenditure of about \$11,000,000, and this sum the committee proposed to raise by loan, and reported a bill for that purpose. The bill was considered in the committee of the whole February 24, and seems to have caused but little debate, passing to its third reading by yeas 92, nays 29.

— *Bayley, ch. 2 — the legislative process by which the Loan was authorized in March 1812 (declaration of war followed in June).*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Under these acts the sum of \$8,134,700 was obtained, reimbursable as provided for in the act, and also a temporary loan for \$2,150,000, reimbursable as provided for in special contracts with different banks. These loans have been entirely repaid.

— *Bayley, ch. 2 — actual amount placed (74% of the \$11M authorization) and the eventual full retirement.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Drift Toward War

The path to the War of 1812 had been a slow accumulation of grievances over British naval policy:

- **Impressment:** The Royal Navy had impressed (forcibly conscripted) thousands of American sailors, claiming they were British deserters.
- **Orders in Council:** British restrictions on neutral trade with French-controlled Europe damaged American commerce.
- **Indian conflicts:** American settlers blamed British support for Indian resistance in the Northwest Territory.
- **The *Chesapeake-Leopard* affair (1807):** A British warship had attacked a U.S. Navy vessel, killing three Americans.

President Madison's war message of June 1, 1812 enumerated these grievances. Congress declared war on June 18, 1812.

### The Financing Problem

What no one had adequately planned for was **how to pay for the war**. Gallatin had warned for years that war finance required either a National Bank (which he supported) or a much larger tax program (which Republicans opposed on ideological grounds). The Republican-dominated Congress had let the First Bank expire in February 1811 — over Gallatin's objections — leaving the Treasury without its primary financial intermediary just months before the war began.

The Six Per Cent Loan of 1812 was launched into this difficult environment. Subscriptions were sluggish. New England banks largely abstained. The Treasury was forced to accept payment in state bank notes (rather than specie), and many of those notes traded below par. The effective price the government realized was therefore below the nominal par at which the bonds were issued.

## Quantity Outstanding

The Loan reached its \$7.81 million peak in mid-1815 — three years after issuance. ## Secondary Market Price

The price history of the Six Per Cent Loan of 1812 is one of the most dramatic in the Hall-Payne-Sargent database. The full range — **\$75 to \$109.50** — represents a 46% peak-to-trough swing over a single decade.

1. **April 1813 (\$95-100):** Initial trading. Despite war uncertainty, prices held near par as patriotic subscribers absorbed the early tranches.

2. **1813–1814 Slow Decline:** Battlefield reverses (especially in the Northwest Territory), British naval pressure on the East Coast, and the burning of Washington (August 1814) pushed prices steadily downward.

3. **January 1815 Trough (\$75):** The lifetime low. By January 1815, the British navy had blockaded most American ports, federal credit was in tatters, and several major financial institutions were close to insolvency. The Treaty of Ghent had been signed in December 1814, but news had not yet reached the United States. This was the **darkest hour** of American sovereign credit since 1789.

4. **Spring 1815 Recovery:** News of the Treaty of Ghent and Andrew Jackson's victory at New Orleans arrived in February. Prices spiked sharply, recovering most of the war-period decline within weeks.

5. **1816–1818 Plateau:** Prices held near par as the Second Bank of the United States (chartered April 1816) restored some financial stability.

6. **1819–1821 Volatility:** The Panic of 1819 — a severe contraction triggered by the Second Bank's curtailment of credit and the collapse of post-war commodity prices — initially weighed on prices, but as the crisis abated, prices climbed to the lifetime peak (\$109.50) by mid-1821, reflecting the bond's status as a high-grade investment in a deflationary environment.

7. **1822–1825 Decline to Par:** As the redemption date approached, prices converged to par.

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20048, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### A Difficult Lesson Learned

The Six Per Cent Loan of 1812's troubled placement experience taught a hard lesson: **wartime sovereign borrowing requires institutional infrastructure**. Without the First Bank of the United States to act as fiscal agent, marketer, and reserve depository, the Treasury found itself struggling to place even modest amounts of debt. This experience directly motivated the **chartering of the Second Bank of the United States in 1816**, which would handle the much larger refundings of the late 1810s and 1820s with vastly greater efficiency.

### The Madison Administration's Fiscal Failure

More broadly, the difficulty of placing this Loan revealed deep flaws in Madisonian war finance. The Republican administration had been ideologically opposed to taxes, central banks, and large standing armies — but war required all three. The Six Per Cent Loan of 1812's troubled history is one of several instances where Republican ideological commitments collided with wartime fiscal necessity. The lesson was absorbed reluctantly, but absorbed: by the late 1810s, even Madison and Monroe accepted institutional financial reform.

### A Surviving Veteran

Despite its rocky start, the Six Per Cent Loan of 1812 ultimately served its purpose. The \$7.81 million it raised — combined with the larger Sixteen Million Loan of 1813 (Bond ID 20052), the Seven and One-Half Million Loan of 1813 (Bond ID 20054), and other instruments — financed essential war operations: army payrolls, naval construction, defense of New Orleans. By the time it was fully retired in the mid-1820s, it had outlived the war that called it into being and witnessed the rebirth of American financial institutions.
