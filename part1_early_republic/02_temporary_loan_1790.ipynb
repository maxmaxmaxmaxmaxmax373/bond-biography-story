## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The mass of debt which devolved upon the United States, as an inheritance from the Continental Congress and the several colonies, engaged the attention of the national legislature soon after the adoption of the Constitution.

— Bayley, *National Loans*, ch. 2, p. 83.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Temporary Loan of 1790** is the second federal short-term financing operation, the direct successor to the Temporary Loan of 1789. Authorized by the **Act of March 26, 1790** — the first congressional statute specifically authorizing federal short-term borrowing — it served the same essential purpose: providing working capital to the Treasury while longer-dated funding mechanisms (the great August 1790 funding operation) were being placed.

What distinguishes the 1790 loan from its 1789 predecessor is its **statutory authority**: rather than being executed under Hamilton's discretionary Secretarial powers, it was authorized by a specific Act of Congress. This was a small but consequential institutional development. By formalizing the Treasury's short-term borrowing in statute, Congress was beginning to construct the legal-administrative architecture that would later support large-scale federal credit operations.

The loan was tiny — peak outstanding of approximately **\$30,000** — and recorded in only one observation in the Hall-Payne-Sargent database. Like its 1789 predecessor, it was placed primarily with the Bank of North America and the Bank of New York, retired within months, and never traded on the secondary market.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> The seventh section of the act of March 26, 1790 (1 Statutes, 105), authorized the President to empower the Secretary of the Treasury to make such loans as might be requisite to carry into effect certain specific appropriations made by the act.

— *Bayley, ch. 2 (Temporary Loan of 1790), p. 62.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> That, in order to be able to furnish, in the course of the ensuing month, for the compensation of the members of Congress and the officers and servants of the two houses, a sum of about \$60,000; for the payment of the salaries of the civil list to the end of the present month, a sum of about \$40,000; for the use of the Department of War, a sum of about \$50,000; and for procuring bills to pay an arrear of interest on the Dutch loans to the first of June next, a sum of about \$35,000 (amounting together to about \$185,000) — it will be requisite to obtain a loan of \$100,000.

— *Hamilton to the President, March 29, 1790 — quoted in Bayley, ch. 2 — itemizing the immediate expenditures the loan was to bridge.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> The Secretary of the Treasury is hereby authorized to negotiate and agree for a loan to the United States, to an amount not exceeding \$100,000, bearing an interest not exceeding 6 per cent. per annum.

— *George Washington, executive authorization, March 31, 1790 — quoted in Bayley, ch. 2.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Act of March 26, 1790

The Act of March 26, 1790 was a procedural statute, modest in size and scope. It authorized the Secretary of the Treasury to borrow such sums as might be necessary on the credit of the United States, to be repaid from forthcoming customs revenues. The key innovations:

1. **Statutory authority:** Unlike the 1789 loan, which had been executed under inherent executive authority, the 1790 loan was now grounded in a specific congressional statute. This anticipated the later constitutional norm that all federal borrowing must be authorized by Congress.

2. **Repayment provisions:** The Act specified that repayment would come from customs revenues, establishing the principle that **specific revenue streams could be pledged to specific debt obligations** — a principle that would later be greatly elaborated in the Funding Act of August 1790.

3. **No fixed term:** The loan was authorized as "temporary," with no specific maturity date. This gave Hamilton operational flexibility to roll the debt as needed during the spring and summer of 1790.

### Context: Hamilton's Funding Plan

The Act of March 26, 1790 came at a critical moment. Hamilton's *Report on Public Credit* of January 1790 had proposed a sweeping plan to refund all Revolutionary War debt at par, with a mix of 6%, 3%, and deferred-6% coupons. The plan was facing fierce political opposition (especially from James Madison, who advocated discrimination between original and current bondholders), and its passage was uncertain through the spring and early summer of 1790.

The Temporary Loan of 1790 was therefore the **bridge financing** that kept the Treasury operational while the political battle over the larger funding plan played out. It was retired in late 1790 once the Funding Act of August 4 had been enacted and the new long-term Stocks were being issued.

```python
fig, ax = plt.subplots(figsize=(16, 4.5))
from datetime import datetime as dt

lifecycle = [
    ('Hamilton Report', '1790-01-09', '#2ecc71'),
    ('Authorizing Act', '1790-03-26', '#2ecc71'),
    ('Compromise of 1790', '1790-07-26', '#f39c12'),
    ('Funding Act', '1790-08-04', '#2ecc71'),
    ('First Issue', '1790-12-31', '#1f77b4'),
    ('First BUS Chartered', '1791-02-25', '#2ecc71'),
]


for i, (label, date_str, color) in enumerate(lifecycle):
    d = dt.strptime(date_str, '%Y-%m-%d')
    y_pos = 0.3 if i % 2 == 0 else -0.3
    ax.plot(mdates.date2num(d), 0, 'D', color=color, markersize=9, zorder=5)
    ax.annotate(f'{label}\n{date_str}', (mdates.date2num(d), y_pos),
                ha='center', va='bottom' if y_pos > 0 else 'top',
                fontsize=7.5, fontweight='bold', color=color,
                arrowprops=dict(arrowstyle='->', color=color, lw=0.8))

ax.set_yticks([])
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_major_locator(mdates.YearLocator(2))
ax.set_title('Temporary Loan of 1790 — Lifecycle Timeline', fontsize=14, fontweight='bold')
ax.set_ylim(-1.2, 1.2)
ax.legend(loc='upper right', fontsize=9)
plt.tight_layout()
plt.show()
```

## Legacy and Significance

### Statutory Codification of Treasury Borrowing

While the 1789 Temporary Loan demonstrated that the federal government *could* borrow, the 1790 Temporary Loan established the principle that **federal borrowing must be authorized by Congress**. This is now so fundamental to American fiscal practice that it is easy to forget how recent its institutionalization is. The Act of March 26, 1790 is, in this sense, a landmark.

### A Pattern That Would Recur

The Temporary Loan of 1790 set the pattern for a series of similar short-term loans (Bond IDs 20025–20033), each authorized under a separate statute and used by the Treasury to bridge gaps between revenue receipts and expenditures. This is the ancestor of the modern Treasury bill — a short-term, specific-purpose borrowing instrument used to manage the Treasury's day-to-day cash flow.

### Hamilton's Operational Genius

Both the 1789 and 1790 Temporary Loans demonstrate Hamilton's operational genius. The big policy battles — *Report on Public Credit*, the Funding Act, the First Bank of the United States, the Excise Tax — get most of the historical attention. But the unsung work of keeping the Treasury operationally functional during a period of intense political controversy was equally important. The Temporary Loans were the instruments through which Hamilton kept the lights on.
