## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of February 11, 1807 (2 Statutes, 411), the Secretary of the Treasury was authorized to exchange the unpaid balances of these stocks for a new stock, bearing 6 per cent. interest, and exempt from redemption until eight years should expire.

— Bayley, ch. 2, on the Exchanged and Converted Stocks of 1807.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Exchanged Six Per Cent Stock of 1807** is one of two paired refunding instruments (alongside the Converted Six Per Cent Stock of 1807, Bond ID 20046) created by the **Act of February 11, 1807**. Together, they were the centerpiece of Treasury Secretary **Albert Gallatin's** systematic effort to **rationalize and consolidate** the diverse mass of older federal debts that had accumulated by the early 1800s.

By 1807, the federal debt portfolio had become bewildering. Holders carried Six Per Cent Stocks (1790, 1796), Three Per Cent Stocks (1790), Deferred Six Per Cent Stocks (1790), Eight Per Cent Loans (1798, 1800), Navy Stocks, Louisiana Six Per Cent Stocks, and various smaller temporary loans. Each had its own maturity, call features, payment schedule, and tax treatment. From the Treasury's standpoint, this complexity was operationally burdensome. From the holder's standpoint, it was confusing and reduced the marketability of older securities.

The Act of February 11, 1807 authorized Gallatin to **exchange older obligations for new, standardized 6% Stocks** with consistent terms. Holders who chose to participate would surrender their old securities and receive in return:
- The **Exchanged Six Per Cent Stock** (this bond, ID 20045) — for direct one-for-one exchange of older 6% obligations into a newly-numbered, current series.
- The **Converted Six Per Cent Stock** (Bond ID 20046) — for conversions of fractional or non-6% holdings into 6% Stock equivalents.

In the Hall-Payne-Sargent database, this Exchanged Stock has **no recorded secondary-market price series** — consistent with its character as a refunding instrument that was largely held by long-term investors rather than actively traded. Quantity peaked at **\$5.99 million** in March 1808 and was retired by the early 1820s.


## Gallatin's Debt-Reduction Program

**Albert Gallatin** (Treasury Secretary 1801–1814) is one of the most underappreciated figures in early American fiscal history. A Geneva-born scholar and long-serving Republican Congressman, Gallatin took office in 1801 with a single overarching goal: **to reduce the federal debt to zero**. He believed (with Jefferson) that excessive sovereign debt corrupted republican government, and he committed himself to systematic debt retirement.

### The Mechanics of Reduction

Gallatin's program had three pillars:

1. **Budget Surpluses:** By cutting military and diplomatic spending and benefiting from rising customs revenues, Gallatin generated consistent fiscal surpluses through the 1801–1812 period.

2. **Sinking Fund Discipline:** He used the Sinking Fund (created by Hamilton in 1790) systematically to retire outstanding debt, applying surpluses to bond purchases or redemptions.

3. **Debt Consolidation:** This is where the Act of February 11, 1807 came in. By consolidating the older debt into standardized instruments, Gallatin made the remaining outstanding debt easier to track, easier to retire, and cheaper to service.

### The Refunding Instruments

The Exchanged Six Per Cent Stock and the Converted Six Per Cent Stock together represented a small but symbolically important step in this program. They allowed Gallatin to:
- Replace older, irregular 6% obligations with a new uniform series.
- Eliminate the operational complexity of tracking multiple older instruments.
- Position the remaining debt for systematic retirement through the Sinking Fund.

### Why "Exchanged" vs. "Converted"

The two stocks served slightly different purposes:
- **Exchanged Stock:** Direct one-for-one exchange of older 6% obligations whose principal amount could be carried into the new instrument unchanged.
- **Converted Stock:** For obligations that required arithmetic conversion (e.g., older instruments with different coupon rates that needed to be re-priced into 6% equivalents).

Together, they handled essentially the entire surviving stock of late-1790s and early-1800s federal debt.

## Quantity Outstanding

The Exchanged Six Per Cent Stock peaked rapidly at \$5.99 million in March 1808 — almost immediately after issuance authority was activated. This pattern is typical of refunding instruments: holders elect to participate quickly (within months), and outstanding balance reaches its peak almost immediately. After 1810, balance declined steadily as Gallatin (and his successors) used Sinking Fund operations to retire the Stock. By 1820, outstanding had fallen to under \$2 million; by 1824, the Stock was effectively extinguished.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1812-06-18'), pd.Timestamp('1815-02-18'), alpha=0.15, color='#e74c3c', label='War of 1812')
ax.axvspan(pd.Timestamp('1819-01-01'), pd.Timestamp('1821-12-31'), alpha=0.15, color='#9b59b6', label='Panic of 1819')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20045, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Exchanged Six Per Cent Stock of 1807 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20045, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.2f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.3f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Price Data

The Hall-Payne-Sargent database does not record a secondary-market price series for this specific Stock. Refunding instruments of this kind were typically held by long-term investors who had elected to participate in the exchange, and they generally traded only in line with the underlying 6% Stock benchmark. For traded-price views of comparable 6% federal securities of the period, see the chapter on the Six Per Cent Stock of 1790 (Bond ID 20021) and the Louisiana Six Per Cent Stock (Bond ID 20044).


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20045, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1812-06-18'), pd.Timestamp('1815-02-18'), alpha=0.15, color='#e74c3c', label='War of 1812')
    ax.axvspan(pd.Timestamp('1819-01-01'), pd.Timestamp('1821-12-31'), alpha=0.15, color='#9b59b6', label='Panic of 1819')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Exchanged Six Per Cent Stock of 1807 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Exchanged Six Per Cent Stock of 1807 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Legacy and Significance

### Standardization Triumph

The Exchanged Stock represents a quiet triumph of **financial standardization**. By collapsing multiple older 6% obligations into a single uniform series, Gallatin reduced the operational complexity of the federal debt and made the Treasury's job easier. This is the same impulse that would later drive the **Consols of 1865** under McCulloch and the **Funding Act of 1870** under Sherman — and it ultimately leads to modern Treasury practice, where outstanding federal debt is divided into a small number of standardized instrument classes (bills, notes, bonds, TIPS).

### Part of the Jeffersonian Debt Reduction

Together with its sister Converted Stock and the broader Sinking Fund operations, the Exchanged Stock helped Gallatin make extraordinary progress in reducing the federal debt. Federal debt outstanding fell from **\$83 million in 1801** (when Gallatin took office) to **\$45 million in 1812** (the eve of war) — a reduction of nearly half over a single decade. This reduction would have been even more dramatic but for the Louisiana Purchase, which added \$11.25 million to the debt.

### The War Interrupts

The War of 1812 ultimately reversed Gallatin's debt-reduction program. By 1816, federal debt had grown back to **\$127 million** — its highest level in absolute terms since 1790. But the institutional architecture Gallatin had built — including the standardized refunding mechanisms exemplified by the Exchanged Stock — would prove instrumental in managing the much larger debt burden of the postwar era.
