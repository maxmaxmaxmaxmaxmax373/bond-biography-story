## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> It will thus be seen that, "in the case of the debt of the United States, interest upon two-thirds of the principal only, at 6 per cent., was immediately paid; interest upon the remaining third was deferred for ten years, and only 3 per cent. was allowed upon the arrears of interest, making one-third of the whole debt."

— Bayley, ch. 2, summarizing Hamilton's *Report on the Public Credit*, p. 86.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Deferred Six Per Cent Stock of 1790** is one of the most ingenious pieces of financial engineering in American history. Created by the **Funding Act of August 4, 1790**, alongside the regular Six Per Cent Stock and the Three Per Cent Stock, the Deferred Stock had one extraordinary feature: **it paid no interest for the first ten years**. Its 6% coupon began accruing only on **January 1, 1801**.

This was Hamilton's solution to a difficult fiscal arithmetic. Holders of Revolutionary War debt expected to be paid at par. The total face value of the old debt to be funded was approximately \$77 million. Paying 6% on the entire \$77 million immediately would have cost the new federal government \$4.6 million per year in interest — far more than its early customs revenues could support. Yet defaulting on the bonds, even partially, would have destroyed American credit before it began.

Hamilton's solution: **issue par-value bonds, but stagger the coupon-payment dates**. For every \$100 of old principal, holders received \$66.67 of regular Six Per Cent Stock (paying interest immediately) and \$33.33 of Deferred Six Per Cent Stock (paying interest only after 1801). The face value was honored at par, satisfying the political imperative. The cash-flow burden was effectively reduced by one-third for the first decade, satisfying the fiscal imperative.

In the Hall-Payne-Sargent database, the Deferred Stock has **381 monthly price observations** (December 1790 – December 1823). Mean price: **\$89.03**. Range: **\$45.00** (December 1790, immediately after issuance, when the deferred-coupon discount was sharpest) to **\$104.00** (November 1809, after the deferred coupon had been paying for nearly nine years). Peak outstanding: **\$13.69 million** (December 1797).


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> Section 15 provides: "That for two-thirds of any sum subscribed to the said loan, by any person or persons, or body politic, which shall be paid in the principal and interest of the certificates or notes issued as aforesaid by the respective states, the subscriber or subscribers shall be entitled to a certificate, purporting that the United States owes to the holder or holders thereof, or his, her, or their assigns, a sum to be expressed therein, equal to two-thirds of the aforesaid two-thirds, bearing an interest of six per centum per annum, payable quarter-yearly, and subject to redemption by payments, not exceeding in one year, on account both of principal and interest, the proportion of eight dollars upon a hundred of the sum mentioned in such certificates; and to another certificate, purporting that the United States owe to the holder or holders thereof, his, her, or their assigns, a sum to be expressed therein, equal to the proportion of thirty-three dollars and one-third of a dollar upon a hundred of the said two-thirds of such sum so subscribed, which after the year 1800 shall bear an interest of six per centum per annum, payable quarter-yearly, and subject to redemption by payment of the sum specified therein whenever provision shall be made by law for that purpose."

— *Bayley, ch. 2, quoting Section 15 of the Funding Act of August 4, 1790 — establishing the deferred-coupon mechanics.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> It appeared by the report of Albert Gallatin, Secretary of the Treasury, January 20, 1806, that there remained outstanding: of the old 6 per cent. and deferred stocks about \$31,800,000, and of the 3 per cents (nominal value) about \$19,050,000 — total, \$50,850,000. Of these stocks the 6 per cent. and deferred were redeemable at the rate of 2 per cent. per annum, each holder receiving, on every \$100 worth of stock, \$6 for interest and \$2 as reimbursement. This mode of payment had in effect, though not in terms, converted the stocks into annuities, terminable in from twelve to eighteen years later. Under this arrangement these stocks sold in England, where a very large proportion of them was held, at about 95 per cent. on the unredeemed amount of each certificate.

— *Bayley, ch. 2 — Gallatin's January 1806 inventory of outstanding 1790 Stocks, including the 95% London quotation.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Mr. Gallatin proposed that, in exchange for these three stocks, a common 6 per cent. stock should be offered to the public creditors; the old 6 per cent. and deferred stocks to be received at the par value of the unredeemed amount, and the 3 per cents at about 60 per cent. Mr. Gallatin thought the exchange would be advantageous to the creditors, because, 1st, instead of receiving, as they did, quarter-yearly, a small reimbursement of their capital, which was in effect a long annuity, they would, if they exchanged their stocks, receive the whole amount within a much shorter time; and, 2d, because an annuity for a term of years is always worth less in market than its intrinsic or arithmetical value.

— *Bayley, ch. 2 — Gallatin's policy reasoning for the 1807 consolidation that ultimately retired the Deferred Stock.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Mathematics of Deferral

The Deferred Stock's discount at issuance reflects basic bond mathematics. A 6% coupon bond paying immediately is worth more than a 6% coupon bond that begins paying ten years from now — even if both are otherwise identical and both have par face value. At a contemporary discount rate of, say, 6%, a \$100 bond whose coupon stream is delayed by 10 years would be worth approximately:

- **Present value of deferred coupon stream (years 11–perpetuity, at 6%):** ~\$56
- **Present value of optional principal redemption:** marginal addition

So the Deferred Stock should rationally have traded at roughly \$56 immediately after issuance — and indeed, the Hall-Payne-Sargent database shows it opened at about \$45-50 in late 1790. As the deferral period shortened, the price climbed: by 1800 (one year before the coupon activated), the Deferred was trading near par. The **time-decay of the deferral discount** is one of the cleanest natural experiments in early-American bond pricing.

### A Pioneering Use of Deferred Coupons

The Deferred Six Per Cent Stock was an unusual instrument by 18th-century standards. Sovereign debt with deferred coupons was not unknown — the British had used similar techniques during the Napoleonic Wars — but the systematic use of deferral as a debt-restructuring tool was a distinctly Hamiltonian innovation. It allowed the Treasury to honor its obligations without crushing its early cash flow.

### A Lesson for Later Refundings

The Deferred Stock's success would inform later American debt-restructuring operations. Variants of the deferral technique appear in:
- The **Exchanged Six Per Cent Stock of 1807** (Bond ID 20045), which converted older debts into new instruments with revised payment streams.
- The **post-Civil War refunding bonds** under the Funding Act of July 14, 1870.
- 20th-century debt exchanges and Treasury refundings.

## Secondary Market Price

The Deferred Stock's price history is one of the most analytically interesting in the database. The dominant feature is the **systematic premium-build over time** as the deferred coupon activation date approaches. From an initial low of \$45 in late 1790, prices climbed steadily through the 1790s as the time-to-coupon shortened. By 1800, the Stock was trading near par. After January 1, 1801 — when the coupon activated — the Stock traded essentially as a regular 6% bond, with prices moving in line with the Six Per Cent Stock.

The peak price (\$104 in November 1809) reflects the same factors that drove the Six Per Cent Stock to similar levels: confidence in the federal government's credit, low domestic interest rates, and the absence (yet) of major foreign threats. After the War of 1812, the Stock declined modestly but continued trading at solid premium until its retirement.

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20022, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### A Pioneering Debt-Restructuring Tool

The Deferred Six Per Cent Stock demonstrated that **sovereign debt could be restructured to ease cash-flow burdens without nominal default**. By honoring face value while staggering coupon activation, Hamilton found a path between the politically impossible (default or partial repudiation) and the fiscally impossible (full immediate coupons on a \$77M debt stock). This template — par-value preservation with cash-flow rearrangement — has been used by sovereign debt managers ever since.

### A Natural Experiment in Bond Pricing

For modern financial historians, the Deferred Stock provides one of the cleanest natural experiments in early-American bond pricing. Because the bond's only difference from the regular Six Per Cent Stock is the timing of coupon activation, the price differential between the two reveals the **implied discount rate the market was using** at any given time. Studies of these spreads (notably by Sylla, Wilson, and Wright) have shown that early-1790s American capital markets were already remarkably efficient at pricing inter-temporal differences in cash flows.

### A Hamilton Innovation

The Deferred Stock is a quiet but distinct Hamilton innovation. It is less famous than the First Bank of the United States or the Mint, but in its own way equally consequential: it solved a fiscal-political dilemma that could have torpedoed the entire funding system. The discount at which the Deferred opened — and its subsequent climb to par by 1800 — is a small monument to financial engineering at the dawn of the republic.
