## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> It will thus be seen that, "in the case of the debt of the United States, interest upon two-thirds of the principal only, at 6 per cent., was immediately paid; interest upon the remaining third was deferred for ten years, and only 3 per cent. was allowed upon the arrears of interest, making one-third of the whole debt."

— Bayley, ch. 2, summarizing Hamilton's *Report on the Public Credit*, p. 86.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Six Per Cent Stock of 1790** is the centerpiece of Alexander Hamilton's funding system — the largest and most consequential of the three bond classes created by the **Funding Act of August 4, 1790** (the others being the Three Per Cent Stock and the Deferred Six Per Cent Stock). Together, these three Stocks consolidated the entire Revolutionary War debt of the United States into standardized, marketable, federally-issued securities. They are, collectively, the **founding instruments of American public credit**.

The Six Per Cent Stock paid a 6% coupon, quarterly, in coin. It was perpetual — i.e., had no fixed maturity — but was redeemable at the government's option. Holders received this bond in exchange for two-thirds of the principal value of their old continental and state debt holdings. The remaining third was issued as Deferred Six Per Cent Stock (which would not begin paying interest until 1801) and Three Per Cent Stock. The conversion ratio was 100:100 (par for par), but with a critical caveat: the coupon on the surrendered debt had often been 6% as well, and back interest was capitalized into the new principal.

In the Hall-Payne-Sargent database, the Six Per Cent Stock has one of the **richest price histories of any 18th-century federal security**: **382 monthly observations** from November 1790 to December 1823. Mean price: **\$95.72**. Range: **\$70.00** (November 1790, immediately after issuance, before market confidence had fully developed) to **\$128.75** (January 1792, during the speculative bubble that culminated in the Panic of 1792). Peak outstanding: **\$28.17 million** (December 1794).


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> Section 2 of this act authorized a loan of \$12,000,000, to be applied to the payment of the principal and interest of the foreign debt: Provided, That no contract should be entered into which should preclude the United States from reimbursing the loan within fifteen years. Section 3 authorized a loan to the full amount of the domestic debt, payable in certificates issued for the said debt, according to their specie value, and computing the interest upon such as bore interest to December 31, 1791. Section 13 authorized a loan of \$21,500,000, payable in the principal and interest of the certificates or notes which prior to January 1, 1790, were issued by the respective states as evidences of indebtedness incurred by them for the expenses of the late war.

— *Bayley, ch. 2, summarizing Sections 2, 3, and 13 of the Funding Act of August 4, 1790 (1 Statutes, 138).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> It will thus be seen that, "in the case of the debt of the United States, interest upon two-thirds of the principal only, at 6 per cent., was immediately paid; interest upon the remaining third was deferred for ten years, and only 3 per cent. was allowed upon the arrears of interest, making one-third of the whole debt. In the case of the separate debts of the states, interest upon four-ninths only of the entire sum was immediately paid; interest upon two-ninths was deferred for ten years, and only 3 per cent. allowed on three-ninths."

— *Bayley, ch. 2, quoting Hamilton's *Report on the Public Credit*.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> The records show that on June 30, 1803, there was standing upon the books of the treasury and the several commissioners of loans, in stock of the United States, \$70,164,781 48, which amount was held and owned by 14,326 parties, including \$32,119,211 88 held by foreigners and \$38,035,569 60 held by states, banks, and other corporations, and by citizens of the United States.

— *Bayley, ch. 2 — June 30, 1803 holdings statement, broken down by foreign vs. domestic.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

### Political and fiscal context

Bayley also frequently records the political backdrop — the wartime urgency, the legislative debate, the economic conditions — that shaped the placement:

> The dividend books relative to this stock were deposited in and paid at the office of the Bank of the United States, at Washington, to the proprietors, that part excepted which was payable to stockholders (principally foreigners) who had given permanent powers of attorney to the Bank of the United States, and which was remitted to that institution at Philadelphia.

— *Bayley, ch. 2 — describing the operational mechanics of paying interest on the 1790 Stocks.*

Read the modern paragraphs that follow with these four Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Funding Act of August 4, 1790

The Funding Act of August 4, 1790 — formally "An Act making provision for the (payment of the) Debt of the United States" — is one of the most consequential statutes in American history. Drafted by Hamilton, debated for months in Congress, and finally enacted only after the famous **Compromise of 1790** (in which southern Republicans accepted federal assumption of state war debts in exchange for relocating the capital to the Potomac), it transformed American public credit virtually overnight.

### The Three Stocks

The Act authorized three classes of new federal bonds, into which holders of old Revolutionary debt could exchange their securities:

| Class | Coupon | Notes |
|-------|--------|-------|
| **Six Per Cent Stock** (this bond) | **6.0% from issuance** | Two-thirds of principal received this |
| **Deferred Six Per Cent Stock** (Bond ID 20022) | **6.0% but interest deferred until 1801** | One-third of principal received this |
| **Three Per Cent Stock** (Bond ID 20023) | **3.0% from issuance** | Used for capitalized arrears of interest |

### The Mechanics of Refunding

Holders of old Revolutionary continentals and state war debt could exchange them at par. For every \$100 of old principal:
- \$66.67 became new Six Per Cent Stock (paying 6% from 1790)
- \$33.33 became new Deferred Six Per Cent Stock (paying 6% from 1801)

Separately, all back interest on the old debt was capitalized — converted into principal — and issued as Three Per Cent Stock.

### The Assumption

A separate provision of the Act authorized the federal government to assume up to \$21.5 million of state war debts. This was the most politically charged element of the package. Madison and Jefferson opposed it; Hamilton secured passage through the famous dinner-table compromise. The assumption mattered both fiscally (it transferred state debts to the federal balance sheet) and politically (it shifted the locus of post-Revolutionary creditor power from the states to the federal government).

### Why It Worked

The Funding Act was a triumph of **financial engineering disguised as moral commitment**. Hamilton's stated principle was that the United States must honor its debts at par — but the *mechanics* of doing so involved subtle reductions in the effective coupon rate (because of the Deferred Stock and the 3% Stock) that made the actual interest burden manageable. Investors received par-value bonds, but the discounted cash flow was actually less than 6% across the entire stock package. This was Hamilton's central insight: **honoring debt at par was politically essential, but the new bonds didn't have to all carry full coupons immediately**.

The Six Per Cent Stock's outstanding balance climbed steadily through the early 1790s as old Revolutionary debts were exchanged. ## Secondary Market Price

The price history of the Six Per Cent Stock is one of the longest and most informative in early-American financial history. Several phases stand out:

1. **November 1790 (\$70):** Initial discount at issuance. Investors were uncertain about the new federal government's commitment to honoring its obligations — and the Stock was, after all, an unprecedented instrument.

2. **1791–early 1792 (\$120+):** Speculative boom. The chartering of the First Bank of the United States (February 1791) and the establishment of the New York and Philadelphia stock exchanges created a vibrant secondary market for the Stocks. Prices soared, peaking at \$128.75 in January 1792.

3. **March–May 1792 (Panic of 1792):** William Duer's speculative empire collapsed in early March 1792, triggering the first significant financial crisis of the new republic. Prices crashed sharply. Hamilton intervened with what is now recognized as the **first U.S. central-bank-style market intervention**, using sinking-fund authority to purchase distressed securities and stabilize prices.

4. **Mid-1790s through Quasi-War:** Prices fluctuated but generally held near par. The Whiskey Rebellion (1794) and the Quasi-War with France (1798–1800) created modest volatility.

5. **Jefferson era (1801–1809):** Prices stayed near par or slightly below. Gallatin's debt-retirement program reduced outstanding balances steadily.

6. **War of 1812:** Modest discount during the war, with prices recovering after Ghent.

7. **Panic of 1819:** Sharp but brief decline as the post-war land-and-credit bubble burst.

8. **Final years (1820s):** Gradual decline as the bonds were retired through redemption.

## Legacy and Significance

### Birth of American Public Credit

The Six Per Cent Stock of 1790 is, with its sister Three Per Cent and Deferred Six Per Cent Stocks, the **single most important debt restructuring in American history**. By converting a chaotic mass of defaulted Revolutionary debt into standardized, marketable, federally-guaranteed securities, Hamilton created the foundation of American public credit. By the mid-1790s, U.S. Stocks were trading at par or above, and **U.S. sovereign debt was considered among the safer investments in the world** — an extraordinary turnaround from the situation just five years earlier.

### Birth of American Securities Markets

The 1790 Stocks were the first widely-traded marketable federal securities. The **New York Stock Exchange** was effectively born around them: the Buttonwood Agreement of May 17, 1792 (named for the buttonwood tree on Wall Street under which traders met) explicitly addressed how the new federal Stocks should be bought and sold. The Philadelphia Stock Exchange (founded 1790) was built around the same instruments. American capital markets, in any modern sense, were born of the 1790 Funding Act.

### Sinking Fund Innovation

The Funding Act also created a **Sinking Fund Commission** (Hamilton, Adams, Jay, Jefferson, Randolph) charged with using budget surpluses to retire outstanding debt. This was an institutional innovation — among the earliest formal sinking funds for sovereign debt anywhere in the world — and it gave the Treasury a tool for systematic debt reduction that would be used for decades.

### Hamilton's Long Shadow

Every subsequent American long-term federal bond — from the Louisiana Six Per Cent Stock (1803) through the Liberty Loans of WWI — owes a debt of design to the Six Per Cent Stock of 1790. The combination of standardized terms, par-value pricing, government redeemability, and reliance on a deep secondary market for liquidity is Hamilton's template, refined and extended over 200+ years.
