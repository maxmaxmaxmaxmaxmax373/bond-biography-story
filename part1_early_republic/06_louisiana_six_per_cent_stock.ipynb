## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The treaty for the cession of Louisiana to the United States was concluded at Paris April 30, 1803.

— Bayley, ch. 2, on the Louisiana Six Per Cent Stock.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Louisiana Six Per Cent Stock** is the bond that financed the **Louisiana Purchase** — the largest land transaction in human history measured by area, and arguably the most consequential single transaction in American history. Authorized by the **Act of November 10, 1803**, it was issued specifically to pay France for the **828,000 square miles** of territory acquired under the Treaty of Cession of April 30, 1803.

The financial transaction was complex. The agreed purchase price was **\$15 million**:
- **\$11.25 million** paid to France in the form of this Louisiana Six Per Cent Stock — a 15-year, 6% bond, redeemable in 1819.
- **\$3.75 million** allocated to U.S. citizens with claims against France (the so-called "spoliation claims" arising from French seizures of American shipping during the Napoleonic Wars).

The Stock was issued at par to France's designated agent (the **Hope & Baring banking syndicate** of Amsterdam and London), who in turn distributed it to European investors. This made the Louisiana Stock the first major American sovereign bond explicitly designed for **international placement** — a watershed moment in the integration of the new American republic into European capital markets.

In the Hall-Payne-Sargent database, the Louisiana Stock has **135 monthly price observations** (July 1809 – September 1820). Mean price: **\$98.14**. Range: **\$75.00** (January 1815, in the closing weeks of the War of 1812) to **\$104.00** (November 1809, the early-Madison-administration peak). Peak outstanding: **\$11.25 million** (March 1804) — the full \$11.25M issuance, paid out almost immediately to France via the Hope & Baring syndicate.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> This loan was contracted to pay France for the province of Louisiana, ceded to the United States by that power April 30, 1803. … The amount was finally fixed at \$15,000,000, of which France was to receive \$11,250,000 in United States bonds, payable in fifteen years, and bearing interest at the rate of 6 per cent. The remainder, amounting to \$3,750,000, was to be devoted to reimbursing American citizens for French depredations on their commerce.

— *Bayley, ch. 2 (Louisiana Six Per Cent. Stock), p. 458.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> The act to issue the stock in payment for the territory, which became known as the Louisiana stock, was approved November 10, 1803 (2 Statutes, 245). It provided that, for the purpose of carrying into effect the convention of April 30, 1803, the Secretary of the Treasury should issue, in favor of the French republic or its assignees, certificates of stock for the sum of \$11,250,000, bearing an interest of six per cent. per annum from the time at which possession of Louisiana might be obtained in conformity with the treaty, the certificates to be delivered by the President to the government of France, or to such persons as should be authorized to receive them, within three months after Louisiana should be taken possession of in the name of the government of the United States.

— *Bayley, ch. 2 — text and operational mechanics of the November 10, 1803 authorizing act.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> The convention provided that the interest should be payable half-yearly, in London, Amsterdam, or Paris, and that the stock should be reimbursed in annual payments of not less than three millions each, the first payment to commence fifteen years after the date of the exchange of ratifications. The annual interest, payable in Europe, was to be paid at the rate of four shillings and sixpence sterling for each dollar payable in London, and of two and a half guilders for each dollar payable in Amsterdam.

— *Bayley, ch. 2 — describing the international payment mechanics of the Louisiana Stock (London/Paris/Amsterdam).*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

### Political and fiscal context

Bayley also frequently records the political backdrop — the wartime urgency, the legislative debate, the economic conditions — that shaped the placement:

> Under this act stock for the portion of the purchase-money due France, amounting to \$11,250,000, was issued. Its redemption began in 1812, and was completed in 1823, every dollar being paid. For the portion reserved to pay American citizens for spoliation (\$3,750,000), no stock was issued, but the claims were paid in money, except the sum of \$11,731 02 carried to the surplus fund June 30, 1868.

— *Bayley, ch. 2 — final resolution of the Louisiana Stock and the spoliation claims.*

Read the modern paragraphs that follow with these four Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Louisiana Purchase

The Louisiana Purchase was, on its face, an unlikely transaction. Napoleon Bonaparte had acquired Louisiana from Spain in 1800 (via the secret Treaty of San Ildefonso). His original plan was to use New Orleans as a launching pad for a renewed French empire in the New World, anchored on a sugar economy in Saint-Domingue (Haiti). The plan collapsed when **the Haitian Revolution** (1791–1804) — the first successful slave revolt in modern history — destroyed the French colonial army on the island, killing tens of thousands and rendering the Louisiana scheme strategically obsolete.

By early 1803, Napoleon needed cash for impending war with Britain and saw little value in retaining a Louisiana he could not defend. President Jefferson, alarmed by reports that Napoleon planned to militarize New Orleans (which would have strangled American commerce on the Mississippi), dispatched **James Monroe** and **Robert Livingston** to negotiate the purchase of New Orleans alone.

To Jefferson's surprise, Napoleon offered the **entire Louisiana Territory** for \$15 million. Monroe and Livingston accepted on April 30, 1803, on instinct and without formal authority. The territory doubled the size of the United States overnight.

### The Constitutional Question

Jefferson, a strict constructionist by philosophy, was deeply troubled by the constitutionality of the purchase. The Constitution did not explicitly authorize the federal government to acquire foreign territory. Jefferson considered seeking a constitutional amendment, but ultimately accepted Treasury Secretary **Albert Gallatin's** advice to proceed under the "implied powers" doctrine. This was an extraordinary political reversal — Jefferson had built his political career attacking implied-powers reasoning during the First Bank debate.

### The Financing Innovation

Gallatin's financing plan was ingenious. Rather than try to raise \$15 million in cash (impossible at the time — federal revenues were under \$11 million per year), the Treasury would **issue \$11.25 million of 15-year bonds directly to the French government**, who would then sell them to European investors via Hope & Baring. The Treasury's only immediate cash outlay was the \$3.75 million for spoliation claims.

This was the first major American sovereign bond placed in **international capital markets**. London and Amsterdam absorbed the Stock readily — the United States was, by 1803, considered a creditworthy sovereign — and Napoleon received approximately \$8.8 million in cash from the European bond syndicate (Hope & Baring kept the difference as their commission).

## Quantity Outstanding

The Louisiana Stock's outstanding balance jumped to its full \$11.25 million in March 1804 — the moment of issuance — and remained there essentially unchanged for several years. Because the bond was issued in a single tranche to the French government (and onward to Hope & Baring), there was no incremental issuance pattern of the kind seen with marketed bonds. After the War of 1812, the Treasury (under Acting Secretary Crawford and then under William Crawford) began modest retirements, accelerating after the formal redemption window opened in 1819. By the early 1820s, outstanding had fallen sharply.

## Secondary Market Price

The Louisiana Stock's price history is short but extraordinarily revealing. It begins in July 1809 — five years after issuance — when the European secondary market for the Stock had fully developed and quotes were being reported regularly in American financial gazettes.

1. **November 1809 (\$104):** Peak. The Madison administration had just taken office, the European wars had not yet directly threatened American shipping, and the Stock's gold-coupon and sterling-redemption appeal made it attractive to British and Dutch investors.

2. **1810–1812 Steady (\$95-100):** A long plateau of stable trading near par.

3. **War of 1812 (1812-1815):** A sustained discount as British naval blockade and military pressure raised credit risk. The trough at \$75 in January 1815 came just weeks before the Treaty of Ghent — but the news of peace had not yet reached London-tied American markets.

4. **Post-1815 Recovery:** Prices climbed back toward par as the war concluded.

5. **Pull-to-Par (1819–1820):** As redemption approached, prices converged to par.

## Legacy and Significance

### Three Cents Per Acre

At a final cost of about **3 cents per acre**, the Louisiana Purchase was one of the great strategic bargains of human history. The territory acquired included all or part of fifteen modern states: Louisiana, Arkansas, Missouri, Iowa, Oklahoma, Kansas, Nebraska, Minnesota (west of the Mississippi), most of Wyoming, Montana, North Dakota, South Dakota, Colorado (east of the Continental Divide), and parts of Texas and New Mexico. The economic, demographic, and strategic implications shaped American history for the next two centuries.

### A Bond That Built a Country

In a literal sense, the Louisiana Six Per Cent Stock **bought half the United States**. No other bond in the Hall-Payne-Sargent database can claim such a direct geographical legacy. Every Civil War battlefield west of the Mississippi, every Great Plains agricultural community, every Western mining boom, every Manifest Destiny pamphlet — all are downstream consequences of this 1804 financial transaction.

### International Capital Markets

The Louisiana Stock's placement through Hope & Baring established a critical institutional bridge between American sovereign borrowing and European capital markets. For the next several decades, the Hope & Baring connection — combined with the Bank of Amsterdam's role as a clearing center — would be the **primary channel through which American Treasury debt reached European investors**. The Louisiana Stock thus marks the moment when American sovereign borrowing became truly international.

### A Quiet Hamilton Postscript

Although Hamilton was no longer at the Treasury (he had left in 1795), the Louisiana Stock was, in design, a Hamilton-style instrument: par-value, coin-payable, marketable, with quarterly coupons and a defined redemption window. Gallatin (Jefferson's Treasury Secretary, but a personal admirer of Hamilton's financial system) preserved the Hamilton template even as the Republicans dismantled other parts of the Hamiltonian state. The Louisiana Stock is a quiet monument to the durability of Hamilton's debt-management innovations across partisan transitions.
