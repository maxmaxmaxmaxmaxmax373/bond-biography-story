## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of August 2, 1813 (3 Statutes, 75), the President was authorized to borrow seven millions and a half of dollars at a rate of interest not exceeding 6 per cent.

— Bayley, ch. 2, on the Seven and One-Half Million Loan.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Seven and One-Half Million Loan of 1813** is the third major War of 1812 bond, authorized by the **Act of August 2, 1813** as a follow-on to the Sixteen Million Loan of February 1813. By midsummer 1813, the Madison administration realized that the original \$16 million authorization would not be sufficient to fund the war through the end of 1814, especially given that **British troops were beginning to redeploy from Europe** as Napoleon's empire crumbled. An additional \$7.5 million was authorized, with terms broadly similar to the earlier Loan.

The Loan was issued at a slightly more favorable price (\$88.25) than the Sixteen Million Loan — a marginal improvement reflecting investor familiarity with the bond format and growing confidence that the war would not result in catastrophic defeat. The same circle of major financiers (Girard, Astor, and to a lesser extent the Boston merchant community) again provided the bulk of the subscriptions, though with significantly more participation from established Mid-Atlantic banking interests now that the institutional pattern had been established.

In the Hall-Payne-Sargent database, this Loan **shares its price series with the Sixteen Million Loan of 1813** (both showing 182 observations, identical statistics, May 1813 – June 1828). This is because contemporary financial markets treated the two bonds as essentially **fungible**: same coupon, same call/maturity structure, similar issue date, identical credit. They typically traded at indistinguishable prices, and the Hall-Payne-Sargent database appropriately uses a unified price series for both.

Peak outstanding for this specific Loan was **\$8.50 million** (June 1814) — slightly above the authorized \$7.5 million due to accrued-interest capitalization, similar to the pattern observed with the Sixteen Million Loan.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> The acts authorizing a loan of \$16,000,000 and an issue of treasury notes for \$5,000,000, in the early part of the year 1813, were intended to provide ample means to meet the deficit of \$19,000,000, which was expected to result from the heavy military and naval operations of the year. The acting Secretary of the Treasury, W. Jones, in his report to the Senate, June 3, 1813, estimated the expenditures for the remainder of the year at about \$29,230,000.

— *Bayley, ch. 2 (Seven and One-Half Million Loan), p. 570.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> He seems to have feared that so large an issue of treasury notes might result in their depreciation, and he advised a loan instead. He says that, in the estimate of resources, "the whole sum of \$5,000,000 authorized to be issued in treasury notes is taken as a part of the resources of the present year. But as it is not deemed eligible to increase the amount of treasury notes in circulation, and as three millions only of those authorized by the act of 1812 were issued in that year, it is respectfully suggested that, in lieu of issuing two millions of the five millions authorized by the act of February, 1813, Congress should authorize an additional loan for the same amount."

— *Bayley, ch. 2 — Acting Treasury Secretary William Jones's June 3, 1813 recommendation that produced the August 1813 loan authorization.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> The bill was slightly amended in the Senate, passed both houses July 30, and was approved August 2, 1813 (3 Statutes, 75). It authorized the President to borrow, on the credit of the United States, a sum not exceeding \$7,500,000, reimbursable at the pleasure of the United States at any time after the expiration of twelve years from January 1, 1814. Under this act \$7,500,000 was obtained in cash, and stock to the amount of \$8,498,581 95 was issued therefor, being a discount of 11.75 per cent.

— *Bayley, ch. 2 — final terms: \$7.5M cash raised against \$8.5M par face value, an 11.75% discount almost identical to the Sixteen Million Loan's 12%.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Continuation of the War

By August 1813, the war's strategic picture was a mix of victories and reverses:

### Bright Spots
- **Battle of Lake Erie** (September 10, 1813): Captain Oliver Hazard Perry's stunning victory secured American control of Lake Erie ("We have met the enemy, and they are ours").
- **Battle of the Thames** (October 5, 1813): Tecumseh killed; Northwest Indian alliance broken.
- **Battle of York** (April 27, 1813): American capture and burning of the Upper Canadian capital (now Toronto).

### Dark Clouds
- **Napoleon's reversals in Europe**: Russia (1812), Leipzig (October 1813), and Paris (March 1814) all freed up British military resources.
- **Naval blockade**: The Royal Navy was tightening its grip on American ports.
- **Treasury exhaustion**: The Sixteen Million Loan had been placed only with Girard's heroic intervention; further borrowing was urgent.

### Why the Second Loan

The Act of August 2, 1813 authorized this additional \$7.5 million in part because Gallatin had foreseen — correctly — that **defeating Napoleon would free up Wellington's veterans to invade the United States**. By the spring of 1814, indeed, large numbers of British regulars were being shipped to North America. The burning of Washington in August 1814, the Battle of Plattsburgh in September, and the New Orleans campaign in December–January all required substantial federal expenditure that this Loan helped fund.

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20054, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### Closing the War Loan Book

The Seven and One-Half Million Loan was the last major war-period long-term bond issued before peace was achieved. (The 1814 Loans — the Ten Million and Six Million — extended the financing further but were issued under increasingly desperate circumstances.) Together with the Sixteen Million Loan and the earlier Six Per Cent Loan of 1812, this Loan completed the trio of major War of 1812 long-bonds.

### The Three-Loan Family

The three major War of 1812 long-bonds — the **Six Per Cent Loan of 1812** (Bond ID 20048), the **Sixteen Million Loan of 1813** (Bond ID 20052), and the **Seven and One-Half Million Loan of 1813** (this bond, ID 20054) — are best understood as a **family of related instruments**. They share:
- Identical coupon (6%)
- Similar maturities (12-13 years)
- Coin-payable status
- Quarterly payments
- The same circle of major subscribers (Girard, Astor, Parish + a small set of Mid-Atlantic and New York merchant houses)

Together, they raised approximately **\$31 million** for the Union war effort — modest by Civil War standards, but enormous in the context of an early-republic federal budget that had averaged \$7 million per year in the late 1790s.

### A Foundation for the Postwar Era

Once peace was achieved, the existence of these three War of 1812 bonds provided the federal government with **a stable, well-priced benchmark** of long-term sovereign debt. The Second Bank of the United States, chartered in 1816, used these bonds as the core of its required reserve holdings. Later refinancings — including the Five Per Cent Loan of 1820 (Bond ID 20071), the Five Per Cent Loan of 1821 (Bond ID 20073), and the Four and One-Half Per Cent Loans of 1824 (Bond IDs 20075, 20077) — were all designed to refinance this 6% war debt at lower postwar rates, much as the Funding Act of 1870 would later refinance Civil War 6% debt.
