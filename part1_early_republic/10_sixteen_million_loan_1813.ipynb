## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of February 8, 1813 (2 Statutes, 798), the President was authorized to borrow sixteen millions of dollars, at a rate of interest not exceeding 6 per cent.

— Bayley, ch. 2, on the Sixteen Million Loan of 1813.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Sixteen Million Loan of 1813** is the largest single bond issue of the War of 1812 and one of the most consequential federal borrowings in early American history. Authorized by the **Act of February 8, 1813**, it was Treasury Secretary **Albert Gallatin's** desperate attempt to fund the second year of a war that was going badly. Its placement was a near-disaster — and its eventual success depended on the personal commitment of three private financiers, two of them Federalists who detested the Madison administration.

The Loan was offered at par with a 6% coupon and 13-year term. But the **financial environment of 1813 was hostile**. The First Bank of the United States had expired in 1811, leaving no institutional backbone for large bond placements. Customs revenues had collapsed under the Embargo and now the British blockade. The British navy controlled the Atlantic seaboard. New England banks, dominated by Federalists, refused on political grounds to subscribe to war bonds. Initial subscriptions came in at a fraction of the amount needed.

What saved the Sixteen Million Loan — and arguably the war effort itself — was the personal intervention of three private financiers:

1. **Stephen Girard** of Philadelphia — a French-born banker, the wealthiest man in America, and (despite his European Republican sympathies) a longtime Federalist supporter of Hamilton-style finance. He took **\$8.18 million** of the Loan.
2. **John Jacob Astor** of New York — fur trade magnate, expanding into finance. He took **\$2 million**.
3. **David Parish** of Philadelphia — Hamburg-born banker representing European capital interests. He took **\$1.86 million**.

Together, these three men subscribed for over **75% of the entire Loan**. Without their commitment, the Madison administration would likely have been forced to seek peace on humiliating terms.

The Loan was sold at a substantial discount: **\$88 per \$100 face value** — a 12% haircut. This reflected the difficult market conditions and was a stark departure from the Hamiltonian par-value tradition.

In the Hall-Payne-Sargent database, the Loan has **182 monthly price observations** (May 1813 – June 1828). Mean price: **\$100.19**. Range: **\$75.00** (January 1815) to **\$109.50** (June 1821). Peak outstanding: **\$18.11 million** (December 1813) — slightly above the authorized \$16 million due to capitalized interest accruals.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> It authorized the President to borrow, on the credit of the United States, a sum not exceeding \$16,000,000, no contract for the same to be entered into which should preclude the United States from reimbursing it at any time after January 1, 1826.

— *Bayley, ch. 2 (Sixteen Million Loan of 1813) — Act of February 8, 1813 (2 Statutes, 798).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> The committee of ways and means, January 18, 1813, reported a bill authorizing the issuing of treasury notes for the service of the year 1813; also a bill authorizing a loan of ___ millions of dollars. It appeared from the report of the committee of ways and means that the expenditures for the year 1813 might be estimated at the sum of \$36,090,000. Of this sum about \$26,000,000 would be required for military and naval operations, which were necessarily heavy, the United States being engaged in war with Great Britain. … To meet this the committee proposed the issue of treasury notes to the amount of \$5,000,000 and a new loan of \$16,000,000.

— *Bayley, ch. 2 — the January 1813 fiscal estimate that produced the \$16M loan authorization.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Under this bill \$15,468,800 was obtained on subscriptions and sales of \$100 in stock for \$88 in money, and \$531,200 at par, by giving an annuity of 1½ per cent. for thirteen years in addition. For the money so obtained, certificates of stock for \$18,109,377 43 were issued, bearing interest at 6 per cent.

— *Bayley, ch. 2 — exact terms of the discount placement: \$88 per \$100, plus an unusual annuity supplement on \$531,200 of par-priced subscriptions.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

### Political and fiscal context

Bayley also frequently records the political backdrop — the wartime urgency, the legislative debate, the economic conditions — that shaped the placement:

> The bill for raising the loan was energetically opposed, on the ground that it contained no special provisions, and set apart no special fund for its reimbursement; that the war, for which great expenses were to be incurred, should never have been declared, and now, that the principal reason for its commencement had been removed by events in Europe, its further prosecution was unnecessary.

— *Bayley, ch. 2 — Congressional opposition to the Loan from anti-war Federalists.*

Read the modern paragraphs that follow with these four Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## Girard, Astor, and Parish: A Wartime Trio

The story of the Sixteen Million Loan is, more than any other early-American bond, the story of three individuals whose private wealth saved a public credit.

### Stephen Girard

**Stephen Girard** (1750–1831) was the central figure. Born in Bordeaux, he had built a vast shipping and banking empire from his base in Philadelphia. By 1810, he was the richest man in America. When the First Bank of the United States's charter expired in 1811, Girard purchased the Bank's Philadelphia headquarters and converted it into the **Bank of Stephen Girard** — effectively a private successor to the BUS.

When the Sixteen Million Loan stalled in early 1813, Girard volunteered to take whatever portion the public market would not absorb. His subscription of **\$8.18 million** — over half the entire issue — was placed at the discounted \$88 price, effectively underwriting the war effort. Girard's bank functioned as the Treasury's primary fiscal agent for the duration of the war.

### John Jacob Astor

**John Jacob Astor** (1763–1848), the German-born fur trade magnate, took **\$2 million** of the Loan. Astor had been thinking systematically about diversifying from the fur trade into finance and saw the war loan as both a patriotic duty and a strategic investment. His participation was particularly notable because he was a Republican (politically aligned with the Madison administration), giving the Loan some bipartisan political cover.

### David Parish

**David Parish** (1778–1826), born in Hamburg, represented European banking interests in the United States. His **\$1.86 million** subscription brought important international capital into the deal. Parish had connections to the **Hope & Baring** syndicate (which had handled the Louisiana Purchase Stock placement) and used those connections to distribute portions of his subscription to European investors.

### The Political Significance

Girard and Parish were both broadly Federalist in sympathy. Their participation in financing a war that the Federalists overwhelmingly opposed was a stark exception to the otherwise rigid party alignment. The episode reveals an important truth about early-American politics: **partisan loyalty had limits when national survival was at stake**. The Federalist financiers who saved Madison's war loan represented a strain of Federalism that placed national credit and federal sovereignty above immediate partisan calculation.

After the war, **Girard would help charter the Second Bank of the United States** (1816), serving on its board and continuing his role as a key intermediary between private capital and the federal government.

## Quantity Outstanding

The Sixteen Million Loan reached its peak (\$18.11M) in December 1813 — within months of issuance, reflecting the front-loaded subscription by Girard, Astor, and Parish. ## Secondary Market Price

The Sixteen Million Loan's price history closely parallels that of its older sibling, the Six Per Cent Loan of 1812. Both bonds carried identical coupons and similar structures, so their secondary-market prices moved largely in tandem. The lifetime low (\$75 in January 1815) and high (\$109.50 in June 1821) coincide exactly between the two bonds.

What distinguishes the Sixteen Million Loan is its **discounted issuance price** (\$88 vs. \$100 par). Investors who took the bond at \$88 effectively earned a yield-to-maturity considerably above the nominal 6% coupon — perhaps 7% or more, depending on the assumed reinvestment rate. This made the Loan attractive to long-term holders like Girard and Astor, who were prepared to wait for price appreciation as the war ended and confidence returned.

## Legacy and Significance

### Financial Heroes of a Reluctant War

The Sixteen Million Loan stands as a testament to the role of **private finance in public emergency**. Without Girard, Astor, and Parish, the Madison administration would have faced an impossible choice: surrender or hyperinflation. Instead, three private financiers — at considerable personal financial risk — provided the credit that kept the war effort alive long enough for a negotiated peace.

### The Case for the Second Bank

The trauma of the 1813 placement was a major factor in the **chartering of the Second Bank of the United States** in 1816. Even James Madison, who had vetoed an earlier Second Bank proposal in 1815, came to accept that institutional financial infrastructure was essential to wartime fiscal capacity. The Second Bank's charter passage was ultimately uncontroversial in 1816, in large part because of the Sixteen Million Loan's near-failure.

### A Discounted Precedent

The decision to issue at \$88 rather than par was a **sharp departure from Hamiltonian doctrine**. Hamilton had insisted that federal bonds always be sold at par, both as a matter of credibility and as a matter of mathematical clarity. Gallatin's willingness to sell at a 12% discount in 1813 was a measure of how dire conditions had become — and a precedent that would haunt the Treasury for decades. Subsequent bond issues, particularly during the Civil War, would similarly accept market-clearing discounts when par-value placement proved impossible.

### A Lasting Legacy in Banking

Stephen Girard's role in saving the Sixteen Million Loan helped cement his reputation and built the financial foundation for what would become **Girard College** (founded by his bequest in 1833) and the long-running **Girard Trust Company** (later absorbed into Mellon Bank, now BNY Mellon). Astor's profits from the Loan, combined with his fur and real-estate empire, made him America's first true multimillionaire. Parish's later trajectory was less happy: he committed suicide in 1826 after speculative losses unrelated to American bonds.
