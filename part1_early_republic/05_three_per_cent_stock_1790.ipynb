## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> It will thus be seen that, "in the case of the debt of the United States, interest upon two-thirds of the principal only, at 6 per cent., was immediately paid; interest upon the remaining third was deferred for ten years, and only 3 per cent. was allowed upon the arrears of interest, making one-third of the whole debt."

— Bayley, ch. 2, summarizing Hamilton's *Report on the Public Credit*, p. 86.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Three Per Cent Stock of 1790** is the third element of Hamilton's funding system, and in many ways its most subtle. Created by the **Funding Act of August 4, 1790** alongside the Six Per Cent and Deferred Six Per Cent Stocks, the Three Per Cent Stock was issued to capitalize **arrears of unpaid interest** on the old Revolutionary War debts.

The economic logic was as follows. The old continental and state debts had nominally paid interest, but in practice many coupons had gone unpaid through the financial chaos of the 1780s. By the time the new federal government took over the debt in 1790, large amounts of accrued-but-unpaid interest had accumulated. Rather than ignore this back interest (which would have been politically unacceptable to creditors) or pay it in cash (which the Treasury could not afford), Hamilton chose a third path: **convert the back interest into new principal**, issued as a 3% coupon bond.

The lower 3% coupon (versus 6% for the principal Stocks) reflected several considerations:
- It signaled that the back interest was being honored in nominal terms, but at reduced economic value.
- It reduced the Treasury's annual interest burden.
- It created a third class of federal security with distinct trading characteristics, deepening the market.

The Three Per Cent Stock had the **longest price history** of any 1790 Stock in the Hall-Payne-Sargent database: **488 monthly observations** from December 1790 to November 1832. Mean price: **\$66.39**. Range: **\$44.17** (June 1798, during the Quasi-War) to **\$100.00** (September 1832). Peak outstanding: **\$19.09 million** (December 1804).


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> And that for the remaining third of any sum so subscribed, the subscriber or subscribers shall be entitled to a certificate, purporting that the United States owes to the holder or holders thereof, his, her, or their assigns, a sum to be expressed therein, equal to the said remaining third, bearing an interest of three per cent. per annum, payable quarter-yearly, and subject to redemption by payment of the sum specified therein whenever provision shall be made by law for that purpose.

— *Bayley, ch. 2, quoting Section 15 of the Funding Act of August 4, 1790 — establishing the 3% Stock as the back-interest-capitalization instrument.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> While the 3 per cents, which were not redeemable at any specific time, were worth in market about 60 per cent. of their nominal value. … A conversion of the 3 per cent. stocks, however, could not be so easily effected, they being worth more in market than a 6 per cent. stock producing the same annual amount, and it would therefore be necessary for the United States in converting this stock to make some sacrifices.

— *Bayley, ch. 2 — Gallatin (1806) on the discount at which the 3% Stock traded relative to the 6% Stock.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> On all subscriptions in 3 per cent. stock, certificates equal to 65 per cent. of the principal of the stock subscribed were to be issued, bearing an interest of 6 per cent., payable quarterly, and not reimbursable without the assent of the holder until after the whole of the 8 per cent. and 4½ per cent. stocks, as well as the stock to be issued by virtue of the act in exchange for the old 6 per cent. and deferred stocks, had been redeemed.

— *Bayley, ch. 2 — describing the 1807 conversion ratio of 3% Stock into Converted 6% Stock at 65 per cent. of nominal value.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Logic of a 3% Coupon

Why did Hamilton choose a 3% coupon for the back-interest stock, rather than (say) 4% or 5%? Several factors:

### Economic Substance

A 3% perpetual is mathematically equivalent (under simplifying assumptions) to a smaller principal at a higher coupon. By issuing back interest as a 3% bond, Hamilton was effectively **reducing the present value of those back-interest claims by half** without explicit haircuts. Investors received a par-value bond, but its cash flow was reduced.

### Symbolic Importance

The 3% rate was the standard coupon on **British Consols** — the prevailing benchmark for high-grade sovereign debt in the late 18th century. By matching this coupon, Hamilton signaled that his new federal Three Per Cent Stock should be regarded as comparable to British Consols in quality.

### Political Acceptability

Domestic creditors — many of them politically connected merchants and Revolutionary veterans — accepted the 3% Stock more readily than they would have accepted a haircut on principal. The math was equivalent, but the optics were very different.

### Long Climb to Par

Because the 3% coupon was below the prevailing 6% market rate for high-grade debt, the Three Per Cent Stock traded at a substantial discount for most of its life. At a contemporary 6% discount rate, a 3% perpetual is worth about **50 cents on the dollar** — and indeed, the Stock traded near \$50 for much of its early years.

The remarkable price climb to **\$100 in September 1832** reflects two forces working in tandem: (1) **falling market interest rates** through the 1820s (as American sovereign credit improved and global rates declined), and (2) **the prospect of redemption at par** as the Jacksonian Treasury moved toward fully extinguishing the federal debt — an event achieved in January 1835.

## Quantity Outstanding

The Three Per Cent Stock's outstanding balance climbed steadily through the 1790s as old back-interest was capitalized into the new Stock. The peak (\$19.09M in December 1804) coincides with the completion of the long Hamilton-era issuance plus continued conversions of older instruments. After 1804, the Treasury (under Gallatin) began aggressively retiring the lower-coupon Stocks, and outstanding balance declined steadily through the 1810s, 1820s, and 1830s.

By the **Jacksonian fiscal-extinction** of January 1835 — when President Andrew Jackson formally declared the federal debt fully retired — the Three Per Cent Stock had fallen to negligible levels.

## Secondary Market Price

The Three Per Cent Stock has the **most distinctive and analytically interesting price history** of any bond in the early Hall-Payne-Sargent database. Three features stand out:

1. **Persistent Discount (1790–1820s):** Because the 3% coupon was below the prevailing market rate, the Stock traded at substantial discount throughout its early life — typically in the \$50-70 range. This was rational pricing, not a sign of distress.

2. **The 1798 Trough (\$44.17):** The lifetime low corresponds to the Quasi-War with France (1798–1800), an undeclared naval conflict that briefly raised credit-risk premia on American sovereign debt. The trough is also remarkable as the only major sub-\$50 observation for any 1790 Stock.

3. **The Long Climb to Par (1820s–1832):** As market interest rates fell through the 1820s and the prospect of debt extinguishment became real under Jackson, the Three Per Cent Stock climbed from the \$60-70 range to par by September 1832. The climb is one of the largest sustained price appreciations in early-American bond market history — a 50%+ rise driven by falling rates and improving redemption prospects.

The **price convergence to par** in 1832 is one of the most important data points in early-American sovereign debt history. It shows that American credit had improved so dramatically in the four decades since 1790 that even a 3% coupon was no longer a discount-trading instrument.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20023, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1812-06-18'), pd.Timestamp('1815-02-18'), alpha=0.15, color='#e74c3c', label='War of 1812')
    ax.axvspan(pd.Timestamp('1819-01-01'), pd.Timestamp('1821-12-31'), alpha=0.15, color='#9b59b6', label='Panic of 1819')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Three Per Cent Stock of 1790 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Three Per Cent Stock of 1790 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20023, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### Capitalizing Back Interest

The Three Per Cent Stock was Hamilton's solution to a vexing problem: **what to do with unpaid back interest on old Revolutionary debts**. The choice to capitalize this interest into a new, lower-coupon instrument — rather than ignore it or pay it in cash — was both fiscally pragmatic and politically astute. It became a template for later sovereign-debt operations whenever a government needed to honor accrued obligations without immediate cash payment.

### Benchmark of American Credit Improvement

The Three Per Cent Stock's price history is, in a sense, a **42-year mirror of American sovereign credit**. Its slow climb from \$44 (1798) to \$100 (1832) tracks the improving creditworthiness of the United States across the entire span of the early republic. By the time the bond reached par, the United States had gone from a precarious, credit-doubtful new nation to a debt-free, AAA-equivalent sovereign — an extraordinary achievement.

### The Jacksonian Endgame

The Three Per Cent Stock was, along with its 1790 siblings, one of the bonds **fully retired by the Jacksonian debt extinguishment of January 1835**. This unique event — when the United States literally had no national debt for the only time in its history — was made possible in large part by the systematic retirement of the lower-coupon 1790 Stocks throughout the 1820s and early 1830s. The Three Per Cent Stock's final price observation in late 1832 stands as a kind of monument to that achievement.
