## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of February 11, 1807 (2 Statutes, 411), the Secretary of the Treasury was authorized to exchange the unpaid balances of these stocks for a new stock, bearing 6 per cent. interest, and exempt from redemption until eight years should expire.

— Bayley, ch. 2, on the Exchanged and Converted Stocks of 1807.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Converted Six Per Cent Stock of 1807** is the smaller, more specialized companion to the Exchanged Six Per Cent Stock of 1807 (Bond ID 20045). Both were authorized by the same statute (the **Act of February 11, 1807**) and serve the same broad purpose — consolidating older federal obligations into a standardized 6% instrument — but they handled slightly different categories of underlying debt.

The **Exchanged Stock** handled direct, par-value swaps of older 6% obligations into a new standardized series. The **Converted Stock** handled more complex conversions — older obligations whose face values needed to be **arithmetically restated** to fit into the new 6% framework. This typically involved fractional balances, accrued interest amounts, or older securities at non-6% coupons that needed to be revalued.

Quantitatively, the Converted Stock was much smaller than the Exchanged. Peak outstanding reached only **\$1.86 million** (March 1808), versus the Exchanged Stock's \$5.99 million peak. This reflects the relative sizes of the two refunding categories: most older 6% obligations could be handled through direct exchange, while only a smaller residual required arithmetic conversion.

The Hall-Payne-Sargent database records **no secondary-market price series** for this Stock. With 130 quantity observations from September 1807 through December 1839, the Converted Stock had a very long retirement tail — small residual amounts remained on Treasury books for over three decades after issuance.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> On all subscriptions in 3 per cent. stock, certificates equal to 65 per cent. of the principal of the stock subscribed were to be issued, bearing an interest of 6 per cent., payable quarterly, and not reimbursable without the assent of the holder until after the whole of the 8 per cent. and 4½ per cent. stocks, as well as the stock to be issued by virtue of the act in exchange for the old 6 per cent. and deferred stocks, had been redeemed.

— *Bayley, ch. 2 — terms of the Converted Stock conversion (3% to 6% at 65 per cent. ratio).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> The only part of it on which a division was called was on filling the blank in the third section so as to entitle the holders of the 3 per cent. stock to an amount of the 6 per cent. stock equal to "65" per cent. of the nominal amount they held. On this the vote was yeas 52, nays 27. The bill was read the third time January 16, and was opposed by Mr. Alston, who said he knew of no advantages that would accrue to the United States from the passage of the bill. Were the United States to be benefited by this modification of the present debt, or were they in the least interested in it? It was clear they were not interested in it, for if they were the keen speculator would not come forward to subscribe.

— *Bayley, ch. 2 — the Congressional debate over the 65% conversion ratio.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Mr. Randolph said he would use a comparison that would come home to the simplest understanding. A farmer owes his neighbor a thousand dollars, for which he has given his bond, bearing an interest at 3 per cent. He has the option of paying it off for \$650. Would any provident man hesitate to discharge the thousand dollars by paying \$650, instead of suffering it to remain as a burden on his children or a lien on his estate? Would it not be most to the advantage of the people of the United States to pay off this stock at the rate of 65 per cent. rather than to let it remain as a perpetual encumbrance on the nation?

— *Bayley, ch. 2 — John Randolph's farmer-and-neighbor analogy in defense of the 65% conversion ratio.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## Why Two Refunding Instruments?

The Act of February 11, 1807 created two paired instruments rather than a single combined Stock for several reasons:

### Operational Clarity

By distinguishing **exchanges** (direct par-value swaps) from **conversions** (arithmetic restatements), the Treasury could maintain clearer records of how the underlying debt had been transformed. This mattered for audit purposes, for tracking accrued-interest obligations, and for clarifying the legal title chain on individual bond holdings.

### Investor Choice

The two-Stock structure also gave bondholders more flexibility. A holder with a clean 6% obligation could elect direct exchange (simple, no recalculation). A holder with a complex bundle of older instruments — perhaps including some non-6% obligations and some accrued interest — could elect conversion (which combined everything into a single new par-value bond).

### A Hamilton-Style Touch

The two-Stock pattern echoes Hamilton's approach in 1790, where the original funding scheme used **three** distinct stocks (Six Per Cent, Deferred Six Per Cent, Three Per Cent) to handle different categories of underlying debt. Gallatin's 1807 refunding used a similar logic, though on a smaller scale: distinguish instrument types based on the nature of the underlying obligation.

## Quantity Outstanding

The Converted Stock's quantity profile is more granular than its sister Exchanged Stock — it shows a longer retirement tail, with small residual amounts remaining on Treasury books well into the 1830s. This reflects the more complex nature of the underlying conversions: many of the underlying obligations were small fractional balances that proved harder to fully extinguish through routine Sinking Fund operations.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1812-06-18'), pd.Timestamp('1815-02-18'), alpha=0.15, color='#e74c3c', label='War of 1812')
ax.axvspan(pd.Timestamp('1819-01-01'), pd.Timestamp('1821-12-31'), alpha=0.15, color='#9b59b6', label='Panic of 1819')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20046, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Converted Six Per Cent Stock of 1807 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20046, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.2f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.3f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Legacy and Significance

### A Footnote with Lasting Implications

The Converted Six Per Cent Stock is, like its sister Exchanged Stock, a footnote in the broader story of American sovereign debt. But it has lasting significance as part of **Gallatin's systematic effort to rationalize and reduce the federal debt**. The instruments themselves are obscure; the program of which they were part was foundational.

### A Predecessor of Modern Refunding Operations

The 1807 paired refunding (Exchanged + Converted) anticipates much later debt-management practices. The post-Civil War **Funding Act of July 14, 1870** used a similar multi-instrument refunding structure (5%, 4.5%, 4% bonds for different categories of underlying debt). The 20th-century practice of **Treasury debt exchanges** — where holders of older securities can elect to swap into newer ones — is a direct descendant of the 1807 framework.

### Closing the Hamilton Era

Together, the Exchanged and Converted Stocks of 1807 mark the **closing of the Hamilton era** of federal debt management. They consolidated the various 1790s and early-1800s instruments into a small number of standardized new Stocks, clearing the way for a fresh framework of debt management going forward. The next major chapter — the War of 1812 financing — would build on this consolidated foundation rather than working through the chaotic mass of older instruments.
