## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The mass of debt which devolved upon the United States, as an inheritance from the Continental Congress and the several colonies, engaged the attention of the national legislature soon after the adoption of the Constitution.

— Bayley, *National Loans*, ch. 2, p. 83.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Temporary Loan of 1789** is, by every meaningful measure, the **first federal loan of the United States of America under the Constitution**. Negotiated by **Alexander Hamilton** in his earliest weeks as the first Secretary of the Treasury, it was tiny by later standards — peak outstanding of just **\$100,000** — but its symbolic and operational importance was enormous.

The Constitution had been ratified in 1788. George Washington took office on April 30, 1789. Hamilton was confirmed as Secretary of the Treasury on **September 11, 1789**. Within weeks, before he had even submitted his famous *Report on Public Credit*, Hamilton needed cash. The new federal government had inherited a chaotic mass of Revolutionary War debts (the "domestic debt"), foreign loans from the Dutch and the French, and obligations to soldiers and suppliers — but no working capital for day-to-day operations. Customs revenues from the new tariff act of July 1789 were starting to flow, but they would take months to accumulate.

The **Temporary Loan of 1789** was Hamilton's bridge financing: a small short-term advance, secured by his own signature as Secretary, on which the federal government could meet its first payroll, pay its earliest contractors, and demonstrate to creditors at home and abroad that the new constitutional government could conduct its financial affairs in an orderly manner. The loan came primarily from the **Bank of North America** (Philadelphia) and the **Bank of New York** — the only two commercial banks of consequence in the United States at the time.

It was retired within months. The Hall-Payne-Sargent database records only **3 quantity observations** (September 1789, December 1789, March 1790), peaking at \$100,000 and falling to \$90,000 by the final observation. No secondary-market price series exists — this was not a marketable instrument in any modern sense, but rather a private credit arrangement between the Treasury and a small handful of banks.


## Hamilton's Working-Capital Problem

The summer of 1789 found the federal government in a peculiar position: rich in moral authority, but acutely short of cash. The three predecessor regimes — the Continental Congress (1775–1781), the Confederation Congress (1781–1788), and the new federal government (March–September 1789) — had together accumulated enormous debts, but the new Treasury had inherited essentially no liquid assets.

### What Hamilton Inherited

| Component | Estimated Amount | Status |
|-----------|------------------|--------|
| Foreign debt (mostly Dutch and French loans) | \$11.7 million | In default on most coupons |
| Domestic debt (continental + state war debts) | ~\$60 million | Trading at deep discounts (15-25¢ on the dollar) |
| Soldier and supplier arrears | uncertain | Largely unpaid |
| Operating cash | ~\$0 | Essentially nil |

### The Tariff of July 4, 1789

The First Congress had passed the Tariff Act of July 4, 1789 — by intentional symbolism enacted on Independence Day — to provide the new government with revenue. But customs duties needed time to be collected, and the receiving infrastructure (customs collectors, ports of entry, weights and measures officials) had to be built from scratch. **In the meantime, the government needed to function**.

### The Loan

Hamilton turned to two banks: the **Bank of North America** (chartered 1781 by Congress, by 1789 a Pennsylvania-chartered private bank) and the **Bank of New York** (founded 1784, partly by Hamilton himself). Both extended short-term credit at 6% interest — a competitive rate for the period — with no formal collateral beyond the Secretary's pledge of forthcoming customs revenues.

This was, in effect, the **first United States Treasury borrowing operation under the Constitution**. The Act of August 4, 1790 (the Funding Act, see Bond IDs 20021–20023) would soon supersede it with a much larger and more ambitious refunding program. But the Temporary Loan of 1789 had already done its essential work: it had proven that the new government could borrow.

## Quantity Outstanding

The database records only three observations of this bond, all in the autumn of 1789 and winter of 1789-1790. Peak outstanding was \$100,000, retiring to \$90,000 within six months. This brief, vanishingly small footprint is faithful to the bond's nature: it was a one-time bridge facility, not a sustained financing instrument.


```python
fig, ax = plt.subplots(figsize=(14, 6))


for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20019, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Temporary Loan of 1789 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20019, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.2f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.3f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Legacy and Significance

### The First Federal Bond

The Temporary Loan of 1789 is, in any meaningful sense, **the first federal bond of the United States under the Constitution**. The earlier Holland Loans (e.g., Bond ID 20009) were initiated under the Articles of Confederation, even if some were managed by the new federal Treasury. Everything else that follows in the Hall-Payne-Sargent database — every Liberty Bond, every Treasury bill, every Five-Twenty — descends from this small autumn-1789 transaction.

### A Demonstration of Constitutional Authority

By executing a borrowing operation in his earliest weeks, Hamilton demonstrated that **the constitutional government had the operational capacity to conduct sovereign finance**. This mattered enormously: the Anti-Federalist critique of the Constitution had warned that a strong central government would be paralyzed by partisanship and incompetence. Hamilton's quick, efficient borrowing rebutted that prediction.

### Establishing a Banking Relationship

The Bank of North America and the Bank of New York extended this credit not from charity but because they believed they would be repaid. By repaying both quickly and at full value, Hamilton **established the model relationship between the federal government and private banks** — a relationship that would expand dramatically with the chartering of the First Bank of the United States in February 1791.

### A Hint of What Was to Come

The Temporary Loan of 1789 is, in essence, a footnote — but it is the footnote that opens the entire history of American sovereign debt. Within twelve months, the same Hamilton would propose, design, and steer through Congress the largest debt-restructuring operation in early-modern history: the funding of \$77 million of Revolutionary debt at 6%, deferred-6%, and 3% coupons, an operation that would make the United States a credit-worthy nation overnight (see Bond IDs 20021, 20022, 20023).
