## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> It authorized the borrowing of \$600,000,000 for the next fiscal year, upon similar terms as had been granted for previous loans, with the exception that nothing authorized by this bill should be made a legal tender or be issued in smaller denominations than \$50.

— Bayley, ch. 4, on the Act of March 3, 1865.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Five-Twenties of 1865** were the final major long-term bond issue of the Civil War era, authorized under the **Act of March 3, 1865** — the last day of the 38th Congress and the same day Hugh McCulloch was confirmed as Fessenden's successor at Treasury. Structurally, they belonged to the well-established Five-Twenty family: 6% coin-payable coupon, callable after 5 years, maturing in 20.

By March 1865, the war was visibly ending. Sherman had completed his March to the Sea. Lee's army was starving in the Petersburg trenches. The Confederate government was preparing to evacuate Richmond. In this context, the Act of March 3, 1865 authorized further borrowing **partly to cover the final months of wartime expenditures, and partly to provide financing for the enormous post-war demobilization and Reconstruction that was about to begin**.

Unlike the 1862 and June-1864 Five-Twenties, the 1865 issue was placed predominantly **after Appomattox**. By the time its first issues were dated (June 30, 1866), the war had been over for more than a year. These were, in effect, **post-war bonds issued under a wartime authorization** — a legislative oddity that reflected how rapidly the fiscal situation was changing in early 1865.

In the Hall-Payne-Sargent database, the 1865 Five-Twenties appear in **124 monthly price observations** (December 1865 – April 1877). Mean price: **\$113.87**. Range: **\$101.88** (December 1865, early-post-war market) to **\$123.88** (July 1869, post-Public Credit Act). Peak outstanding: **\$203.3 million** — the largest of the three Five-Twenty variants after the 1862 issue.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> It authorized the borrowing of \$600,000,000 for the next fiscal year, upon similar terms as had been granted for previous loans, with the exception that nothing authorized by this bill should be made a legal tender or be issued in smaller denominations than \$50. The annual rate of interest was limited to 6 per cent. if payable in coin, or 7 3/10 per cent. in currency, the bonds issued to be redeemable in not less than five nor more than forty years.

— *Bayley, ch. 4 (Five-Twenties of 1865), p. 358 — Act of March 3, 1865 (13 Statutes, 468).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> The laws authorizing the issue of bonds bearing interest in coin specifically pledged the revenue from customs to the payment of that interest and provided for the collection of those duties in the same currency. In the opinion of the Secretary, that pledge should not be violated. A departure from it could only be vindicated by one of those state necessities which justify a nation in temporarily postponing its obligations in order to preserve the power to discharge them at a future day.

— *Bayley, ch. 4 — Treasury Secretary William Pitt Fessenden, *Annual Report*, December 6, 1864, on the policy decision to maintain coin-payment of interest.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Under the authority conferred by these two acts the following 6 per cent. bonds have been issued: Five-twenties of 1865, \$203,327,250; Consols of 1865, \$332,998,950; Consols of 1867, \$379,618,000; Consols of 1868, \$42,539,350. Total, \$958,483,550.

— *Bayley, ch. 4 — the family of four 6% issues authorized under the Acts of March 3, 1865 and April 12, 1866, totaling nearly a billion dollars in face value.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The McCulloch Era Begins

**Hugh McCulloch** took the oath as Treasury Secretary on March 9, 1865 — just weeks before Lee's surrender. A banker by background (he had headed the Indiana banking system before Chase recruited him to head the Comptroller of the Currency), McCulloch was a **contraction advocate**: he believed the war's paper-money expansion had been necessary but inflationary, and he hoped to wind it back as quickly as political conditions permitted.

This philosophy shaped his handling of the 1865 Five-Twenties:

### Larger Issue, Larger Redemption

McCulloch continued to issue the 1865 Five-Twenties throughout 1865 and 1866, ultimately pushing outstanding balance to over \$200 million. Unlike his predecessors, however, he simultaneously began **retiring other Civil War debt** with budget surpluses — a policy of "contraction" that reduced the total debt burden even as new bonds were issued for specific purposes.

### Resistance from Congress

McCulloch's contractionary policy encountered stiff political resistance. **Congressional Republicans**, facing Reconstruction expenses and the need to maintain Union Army garrisons across the defeated South, opposed rapid debt paydown. The 1866 Contraction Act (which authorized McCulloch to retire greenbacks) was partially repealed in 1868 after inflation-averse Westerners and debtor farmers objected. This political tension shaped the market environment for the 1865 Five-Twenties throughout their early years.

## Secondary Market Price

The price history of the 1865 Five-Twenties parallels that of the 1862 issue but with interesting differences:

1. **December 1865 Low (\$101.88):** Entry point of the series. Immediate post-war uncertainty about redemption terms, combined with McCulloch's contractionary messaging, produced a modest discount.
2. **1866–1868 Climb:** Prices rose steadily as wartime risk premia unwound and McCulloch's credibility with markets grew.
3. **July 1869 Peak (\$123.88):** Identical to the 1862 Five-Twenty peak — unsurprising, since the Public Credit Act treated all 6% coin-payable bonds identically.
4. **1870s Decline:** As refunding progressed and the Panic of 1873 took hold, prices drifted down from their 1869 peak. Even so, prices stayed well above par throughout the series.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20113, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
    ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Five-Twenties of 1865 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Five-Twenties of 1865 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20113, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### The Last War Bond

The 1865 Five-Twenties are, formally, the last major long-term bond issued under Civil War wartime authority. Every subsequent major Treasury bond — the Consols of 1865, 1867, and 1868, and the refunding bonds of 1870–1881 — was issued under peacetime authority or under specific refunding statutes. In this sense, the 1865 Five-Twenties mark the **closing of the Civil War fiscal chapter**.

### Benchmark for Refunding

Because they had a clean 6% coupon, coin-payable, with a well-understood 5-year call, the 1865 Five-Twenties became a **benchmark instrument** in the refunding operations of the 1870s. Treasury officials and private bankers compared new lower-coupon bonds (the 4.5% of 1891, the 4% of 1907, the 5% of 1881) against the 1865 Five-Twenty's effective yield when setting prices and call schedules.

### A Quiet Giant

Despite being the second-largest of the Five-Twenty family (\$203 million peak outstanding), the 1865 issue is less famous than its 1862 predecessor. Jay Cooke did not market it with the same fervor (by then he was focused on railroad finance, which would ultimately destroy him in 1873). The gold-vs-greenback debate fixed public attention on the 1862 issue rather than this one. Yet its contribution to Reconstruction-era Treasury operations was enormous — arguably more important in practice than the 1862 bond's political role.
