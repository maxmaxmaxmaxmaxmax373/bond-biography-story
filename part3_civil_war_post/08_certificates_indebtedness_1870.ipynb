## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The act of July 8, 1870 (16 Statutes, 198), authorized the Secretary of the Treasury to issue certificates of indebtedness, not exceeding \$678,362.41, in part payment of the claims for which the Massachusetts and Maine bill had been reported.

— Bayley, ch. 4, on the Certificates of Indebtedness of 1870.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Certificates of Indebtedness of 1870** represent a distinct and specialized category of federal debt. Unlike the large marketable bonds that dominated Civil War financing (Five-Twenties, Ten-Forties, Consols, Seven-Thirties), Certificates of Indebtedness were **targeted instruments**: they were issued under the **Act of July 8, 1870** specifically to pay public creditors who held judicial or administrative claims against the government.

They carried a modest 4% coupon — substantially below the 6% norm of Civil War long-bonds — and a short 5-year term. Peak outstanding was only **\$700,000** — a tiny sum by any measure, and a dramatic departure from the hundreds-of-millions-dollar issues of the war years. But the small size reflected the bond's narrow purpose: it was not a general-revenue instrument but a **claim-settlement instrument**. Holders were typically recipients of specific legal awards or contractual payments owed by the federal government.

The Hall-Payne-Sargent database records no secondary-market price series for this bond — consistent with its limited distribution and its niche role in federal finance. Total-outstanding data shows 60 observations, with the bond peaking near \$700K and declining to near zero by 1875 as certificates matured.


## The Act of July 8, 1870

The Act of July 8, 1870 was one of a cluster of financial and administrative statutes passed by the Forty-First Congress during the Grant administration. Unlike its famous sibling — the Funding Act of July 14, 1870, which authorized the massive refunding of Civil War 6% debt at lower coupons — the July 8 Act was a narrower, more technical measure.

### Purpose

The Act gave the Treasury authority to settle specific classes of claims by issuing **interest-bearing certificates** rather than cash. This was administratively useful: claims against the government often arose from long-dormant wartime contracts, lawsuits over property seizures, or adjudicated awards by commissions (for example, awards to Northern citizens whose property had been damaged by Confederate forces). The government could, rather than paying immediately in cash, issue a 4% Certificate of Indebtedness, redeemable in 5 years.

### Distinction from General Treasury Notes

Unlike the One Year Notes of 1863 (ID 20105), which were broadly marketed as short-term investment instruments, the 1870 Certificates were **issued to specific claimants** rather than sold on the open market. This targeted distribution explains both the small total outstanding (\$700K peak) and the near-absence of secondary-market trading data.

The certificates' life spans the final years of McCulloch's debt-contraction program and the onset of the Long Depression. Their small size meant they were insulated from most of the macro-market turbulence that affected larger Treasury issues.


```python
fig, ax = plt.subplots(figsize=(16, 4.5))
from datetime import datetime as dt

lifecycle = [
    ('Authorizing Act', '1870-07-08', '#2ecc71'),
    ('Funding Act', '1870-07-14', '#2ecc71'),
    ('First Issue', '1871-06-30', '#1f77b4'),
    ('Panic of 1873', '1873-09-18', '#9b59b6'),
    ('Resumption Act', '1875-01-14', '#2ecc71'),
    ('Payable Date', '1875-09-01', '#e67e22'),
]

ax.axvspan(mdates.date2num(dt.strptime('1873-09-18', '%Y-%m-%d')), mdates.date2num(dt.strptime('1879-03-01', '%Y-%m-%d')), alpha=0.15, color='#9b59b6', label='Long Depression')

for i, (label, date_str, color) in enumerate(lifecycle):
    d = dt.strptime(date_str, '%Y-%m-%d')
    y_pos = 0.3 if i % 2 == 0 else -0.3
    ax.plot(mdates.date2num(d), 0, 'D', color=color, markersize=9, zorder=5)
    ax.annotate(f'{label}\n{date_str}', (mdates.date2num(d), y_pos),
                ha='center', va='bottom' if y_pos > 0 else 'top',
                fontsize=7.5, fontweight='bold', color=color,
                arrowprops=dict(arrowstyle='->', color=color, lw=0.8))

ax.set_yticks([])
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_major_locator(mdates.YearLocator(2))
ax.set_title('Certificates of Indebtedness of 1870 — Lifecycle Timeline', fontsize=14, fontweight='bold')
ax.set_ylim(-1.2, 1.2)
ax.legend(loc='upper right', fontsize=9)
plt.tight_layout()
plt.show()
```

## Quantity Outstanding

The quantity series shows the sharp rise-and-fall pattern typical of short-term instruments issued for specific purposes. Outstanding balance rose through 1871–1873, held near \$700K through 1874, and then declined rapidly as certificates matured in 1875.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20118, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Certificates of Indebtedness of 1870 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20118, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.1f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.2f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Price Data

The Hall-Payne-Sargent database records no secondary-market price series for this bond. Given the specialized distribution (certificates issued directly to claimants rather than sold on the market), this absence is expected rather than an oversight.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20118, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Certificates of Indebtedness of 1870 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Certificates of Indebtedness of 1870 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Legacy and Significance

### The Mechanics of Claim Settlement

The 1870 Certificates exemplify a class of federal instruments that played an important but **under-appreciated role in 19th-century public finance**: settlement of claims. Governments routinely owe specific sums to specific creditors for reasons unrelated to the macro financing of the state — unpaid wartime contracts, adjudicated damages, compensatory awards, subsidies to specific industries. In an era without routine budget appropriations, Congress preferred to authorize debt instruments that could be issued to claimants when and as needed.

### A Minor Chapter in the Funding Act Era

The Certificates of Indebtedness of 1870 are typically treated as a footnote in discussions of the Funding Act period. Most historical attention goes to the 5%, 4.5%, and 4% refunding bonds of 1870–1881 (see Bonds 20119, 20120, 20121), which involved billions of dollars of debt turnover. The Certificates, at under a million dollars outstanding, did not affect the macro picture. But they served their narrow purpose efficiently, with predictable cash-flow profiles that let claimants plan around eventual cash redemption in 1875.

### Related Instruments

The 1870 Certificates share the name "Certificates of Indebtedness" with several other bonds in the Hall-Payne-Sargent database — notably the Certificates of Indebtedness issued during the Civil War under the Act of March 1, 1862 (Bond ID 20103) and the WWI-era Certificates of Indebtedness used by Treasury as cash-management instruments. Despite the shared name, these instruments had quite different operational roles: the 1862 Certificates were short-term bank-held money-market paper; the 1870 Certificates were claim-settlement instruments; the WWI-era Certificates were rolling cash-flow instruments similar to modern Treasury bills. The naming reflects an informal tradition in 19th-century U.S. finance of using "Certificate" for specialized short-term paper.
