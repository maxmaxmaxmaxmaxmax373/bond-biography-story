## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The act of March 3, 1863 (12 Statutes, 709), authorized the issue of treasury notes for not exceeding \$400,000,000, of any denomination not less than ten dollars, redeemable after such periods as might be expressed upon their face, payable in not less than six months nor more than two years.

— Bayley, ch. 4, on the Act of March 3, 1863.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **One Year Notes of 1863** occupy a small but distinctive niche in the Union's wartime financing strategy. Unlike the glamorous Five-Twenties (20-year bonds marketed to the public by Jay Cooke) or the Seven-Thirties (3-year notes convertible into long-dated bonds), these were modest, short-term instruments — classic **liquidity paper**, designed to smooth cash-flow mismatches between tax revenues and wartime expenditures.

Authorized under the **Act of March 3, 1863**, the same comprehensive wartime finance act that created the Loan of 1863, the Two Year Notes, and the Compound Interest Notes, the One Year Notes carried a **5% coupon paid at maturity**. They had no semi-annual coupon payments; instead, the note was redeemed a year after issue for **\$105 per \$100 face value**. This structure made them essentially money-market instruments — closer in spirit to a modern Treasury bill than to a traditional bond.

Their role was to provide the Treasury with **working capital** while longer-dated instruments (the Five-Twenties, the Ten-Forties, the Consols) were being marketed and placed. Peak outstanding reached **\$45.5 million** — small in the context of \$2.7 billion in total Civil War debt, but large enough to keep Treasury disbursements flowing when required.

The Hall-Payne-Sargent database does not record a secondary-market price series for this bond. That is not a database limitation but a reflection of market reality: **short-term at-maturity paper typically traded close to its effective yield with little secondary activity**. Holders bought them, held them to maturity, and cashed them in. No deep, liquid secondary market developed.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> That the Secretary of the Treasury be, and he is hereby, authorized to issue, on the credit of the United States, \$400,000,000 in treasury notes, payable at the pleasure of the United States, or at such time or times, not exceeding three years from date, as may be found most beneficial to the public interest, and bearing interest at a rate not exceeding six per centum per annum, payable at periods expressed on the face of said treasury notes; and the interest on said treasury notes and on certificates of indebtedness and deposit hereafter issued shall be paid in lawful money. … And said treasury notes may be made a legal tender to the same extent as United States notes, for their face value, excluding interest.

— *Bayley, ch. 4, quoting Section 2 of the Act of March 3, 1863 (12 Statutes, 710) — the omnibus statute that authorized this Note plus several sister instruments.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> Under this authority treasury notes were issued amounting to \$44,520,000, redeemable one year from date, and \$166,480,000 redeemable two years from date. These notes, bearing interest at 5 per cent. per annum, were a legal tender for their face value, and are generally known as the "one and two year notes of 1863", according to the time for which they were to bear interest.

— *Bayley, ch. 4 — final issuance amounts: \$44.5M of one-year notes plus \$166.5M of two-year notes, all carrying legal-tender status for face value.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Section 1 of this act is as follows: That the Secretary of the Treasury be, and he is hereby, authorized to borrow from time to time, on the credit of the United States, a sum not exceeding \$300,000,000 for the current fiscal year and \$600,000,000 for the next fiscal year, and to issue therefor coupon or registered bonds, payable at the pleasure of the government after such periods as may be fixed by the Secretary, not less than 10 nor more than 40 years from date, in coin, and of such denominations, not less than \$50, as he may deem expedient.

— *Bayley, ch. 4 — the parallel bond authorization in Section 1 of the same Act of March 3, 1863, providing context for the Notes' role in the larger 1863 financing package.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Act of March 3, 1863

The One Year Notes were one instrument in a six-part fiscal package passed as the **Act of March 3, 1863** — one of the most technically sophisticated wartime finance statutes in U.S. history. Drafted at a moment when Union fortunes were precarious (after Fredericksburg, before Chancellorsville), the Act gave Secretary Chase an unusually flexible toolkit:

| Instrument | Bond ID | Coupon | Term | Intended Audience |
|------------|---------|--------|------|-------------------|
| Loan of 1863 | 20104 | 6.0% | 17 years | Banks, institutional buyers |
| **One Year Notes of 1863** | **20105** | **5.0%** | **1 year** | **Treasury liquidity** |
| Two Year Notes of 1863 | 20106 | 5.0% | 2 years | Banks, intermediate liquidity |
| Compound Interest Notes | 20107 | 6.0% | 3 years | Banks, reserve medium |
| Legal Tender Notes (add'l) | — | — | — | Currency expansion |
| Fractional Currency | 30004 | — | — | Retail payments |

The design philosophy was **term diversification**. Rather than float a single enormous bond, Chase and his Congressional allies created a ladder of instruments with different maturities and features, allowing the Treasury to calibrate financing to changing conditions. If long-dated bond demand softened, short-term notes could pick up the slack. If bank reserves grew tight, compound interest notes (which functioned as a kind of interest-bearing reserve asset) could help.

## The Role of Short-Term Paper in Civil War Finance

The One Year Notes occupied a specific role in what modern finance would call **asset-liability management**. Federal revenues in the 1860s arrived unevenly: customs duties were heavy in spring and early summer, internal revenue taxes (newly enacted) were still being ramped up, and many wartime expenditures required cash disbursements in specific seasons (soldier pay at pay-periods, contractor payments at delivery, etc.).

### Bridging Financing

Long-dated bonds like the Five-Twenties took **months to fully place** — even at Cooke's peak, the marketing-to-settlement lag was non-trivial. In the meantime, the Treasury needed cash. Short-term notes like the One Year Notes provided that bridge financing. Banks, insurance companies, and wealthy individuals would buy them as a near-money substitute, confident that either the notes would be redeemed at maturity or could be rolled into the next issue.

### Non-Coin Coupon

A crucial distinction: unlike the Five-Twenties (whose coupons were coin-payable and thus appreciated during greenback depreciation), the One Year Notes' coupon was paid **in lawful money** — i.e., greenbacks. Their effective real yield was therefore subject to greenback inflation. This helps explain why the market never paid a premium over par for these notes: buyers knew they were earning 5% in depreciating currency.

### Limited Price History

The database records essentially no secondary-market price series for this bond. Public-holdings data shows only **6 observations** — a signal that the instrument was held primarily in institutional portfolios and rarely traded. This contrasts sharply with the Five-Twenties, whose 154 price observations reflect an active and well-quoted market. One Year Notes were the Treasury's equivalent of today's commercial paper: useful, widely held, but not glamorous.

## Quantity Outstanding

The outstanding quantity series shows the characteristic **sharp rise and fall** of a short-term instrument: rapid issuance through late 1864, peak near \$45 million, then rapid redemption as notes matured through 1865. The long-tail public-holdings series in the database captures a few lingering observations, but by early 1866 the bond was essentially fully retired.

## Price Data

The Hall-Payne-Sargent database records no secondary-market price series for this bond. That absence is itself informative: it tells us this instrument behaved as near-money rather than as a traded security. The chart below displays the database state faithfully.

## Legacy and Significance

The One Year Notes of 1863 are not a celebrated bond. They did not inspire Cooke-style marketing campaigns, spawn political movements over their redemption medium, or leave a visible mark on macro-financial history. But they played a **real and important role** in the Union war effort.

### Liquidity Provision

They gave the Treasury access to **working capital on days when the major long-dated issues were still being placed**. In wartime, liquidity can be the difference between paying soldiers on schedule and letting morale slip. The One Year Notes were a quiet but essential tool.

### Template for Later Treasury Notes

The experience with One Year Notes (and their siblings, the Two Year Notes and Compound Interest Notes) shaped Treasury thinking about short-term debt for decades. The modern Treasury bill — born in 1929 as a discount instrument — has clear conceptual ancestry in the short-term at-maturity paper of the Civil War.

### A Reminder of Term Diversification

Most importantly, this bond reminds us that the Union's fiscal triumph was **not just about the Five-Twenties**. The war was paid for through a whole laddered portfolio of instruments — some long, some short, some coin-payable, some greenback-payable, some sold to the masses, some quietly held in bank portfolios. Understanding Civil War finance requires understanding all of them, not just the famous ones.
