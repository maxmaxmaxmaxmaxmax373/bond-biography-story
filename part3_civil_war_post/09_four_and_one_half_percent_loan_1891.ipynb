## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The act of July 14, 1870 (16 Statutes, 272), authorized the issue of bonds, in three classes — fifteen years at four and a half per cent., ten years at five per cent., and thirty years at four per cent.

— Bayley, ch. 4, on the Funded Loans of 1881, 1891, and 1907.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Four and One-Half Percent Loan of 1891** is the clearest embodiment of what the Gilded Age Treasury called **"refunding the public debt"** — the systematic replacement of high-coupon Civil War–era obligations with new, lower-coupon instruments. Authorized under the pivotal **Funding Act of July 14, 1870** but not first issued until 1877 (after the Panic of 1873 delayed earlier placement attempts), the 4.5% Loan was a direct response to a simple but powerful fact: by the late 1870s, the general market rate on long-dated, credit-worthy sovereign debt had fallen well below the 6% carried by Civil War Five-Twenties and Consols.

With a **4.5% coupon paid quarterly** (a departure from the semi-annual pattern of earlier Treasury bonds, and closer to the typical European government-bond standard), the bond was designed to appeal simultaneously to:
- **Domestic trustees and savings institutions**, which valued the steadier cash-flow profile of quarterly coupons
- **European investors** in London, Paris, and Amsterdam, who were now more open to American sovereign debt after the resumption of specie payments (1879) had restored the gold standard
- **National banks**, which continued to hold federal bonds as reserves against their banknote issues

The Hall-Payne-Sargent database records the 4.5% Loan in **181 monthly price observations** (September 1876 – September 1891). Mean price: **\$108.89**. Range: **\$99.50** (September 1891, at the redemption point when refunding was imminent) to **\$116.00** (April 1882, the peak of the bond's premium). Peak outstanding: **\$258.0 million**.


## The Funding Act of 1870 — A New Chapter

The **Act of July 14, 1870** is one of the most consequential fiscal statutes of the post-Civil War era. Its stated goal was to **reduce the annual interest burden of the federal debt** by authorizing the Treasury to issue new bonds at lower coupons to refinance outstanding 6% Civil War obligations. The Act created three distinct refunding instruments:

| Bond ID | Name | Coupon | Term | Role |
|---------|------|--------|------|------|
| 20119 | Five Percent Loan of 1881 | 5.0% | 10 years | Intermediate refunding |
| **20120** | **4.5% Loan of 1891** (this bond) | **4.5%** | **15 years** | **Medium-long refunding** |
| 20121 | Four Percent Loan of 1907 | 4.0% | 30 years | Long-term refunding |

The three instruments formed a **refunding ladder**. Over the 1870s, the Treasury used them opportunistically — issuing whichever was best-suited to prevailing market conditions. The 4.5% Loan served the middle of this ladder, intended for investors who found the 5% too short and the 4% too low-yielding.

### Delayed Launch

The 4.5% Loan was authorized in 1870 but not first issued until 1877 because of three factors:
1. **Lingering gold-vs-greenback uncertainty** discouraged European buyers in the early 1870s.
2. **The Panic of 1873** disrupted bond markets for several years.
3. **Republican political consensus** on refunding had to be rebuilt after the 1874 midterms, in which Democrats took the House.

By 1877 — under Secretary **John Sherman** (yes, the same Sherman whose brother had taken Atlanta, and who had been Jay Cooke's key Congressional ally in the 1860s) — all three conditions had improved. Sherman launched the 4.5% Loan in June 1877 and oversaw its successful placement over the next several years.

## Quantity Outstanding

The quantity profile of the 4.5% Loan has a distinctive **triangular shape**:
1. **Rapid issuance 1877–1881** took outstanding balance from zero to its \$258 million peak.
2. **Plateau through the 1880s** as the bond served as a stable benchmark in the refunded debt portfolio.
3. **Sharp decline after 1891** as the redemption date passed and the Treasury called in the bonds (ultimately replacing them with the 2% Funded Loan of 1891 — Bond ID 20127 — which continued the rate-compression cycle).

By late 1891, outstanding had collapsed from \$258M to \$28M, and the bond was essentially retired shortly thereafter.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20120, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Four and One-Half Percent Loan of 1891 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20120, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.1f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.2f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Secondary Market Price

The 4.5% Loan's price history demonstrates several important features of Gilded Age bond markets:

1. **Stable premium through the early 1880s:** Prices held in a narrow \$110–\$116 range from 1879 through 1887, reflecting the bond's role as a benchmark high-grade sovereign instrument.
2. **Gradual decline from 1887–1891:** As the redemption date approached, prices converged toward par — a classic "pull-to-par" effect observed in bonds that are expected to be called or redeemed.
3. **September 1891 trough (\$99.50):** Prices briefly dipped below par as the imminent refunding overhang weighed on valuations.
4. **Quarterly coupon premium:** The quarterly-coupon structure was attractive to trustees and endowments managing regular disbursement schedules. This likely contributed to the bond's sustained premium over most of its life.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20120, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Four and One-Half Percent Loan of 1891 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Four and One-Half Percent Loan of 1891 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20120, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### The Shift to a Gold-Standard Bond Market

The 4.5% Loan of 1891 was part of a broader transformation of American sovereign debt markets. Before specie resumption (January 1, 1879), U.S. Treasury bonds had carried a **gold-risk premium** reflecting uncertainty about whether the government would actually honor its coin-redemption promises. After resumption, that premium largely disappeared. The 4.5% Loan was among the first major Treasury issues placed **entirely under the restored gold standard** — its quarterly coupons and mid-term maturity made it a natural instrument for European investors now returning to American markets.

### Interest Savings

The economic savings from refunding were substantial. A \$258 million bond at 4.5% pays \$11.61 million in annual interest. The same \$258 million at the old 6% coupon (the Civil War rate) would have paid \$15.48 million — a difference of \$3.87 million per year. Over the 15-year life of the bond, the Treasury saved close to \$60 million in interest costs on this single issue alone — and the 4% Loan of 1907 saved considerably more.

### Quarterly Coupons

The 4.5% Loan was one of the first U.S. Treasury bonds to use **quarterly rather than semi-annual coupons**. This feature, borrowed from European practice, was designed to match the cash-flow needs of savings institutions and trustees. It also made the effective yield calculation more intuitive (annual yield = 4 × quarterly coupon). The quarterly feature was continued in the 4% Loan of 1907 but was not universally adopted — most later Treasury bonds reverted to semi-annual payments. Today, all U.S. Treasury notes and bonds pay semi-annual coupons.
