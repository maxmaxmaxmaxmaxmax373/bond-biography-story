## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of June 30, 1864 (13 Statutes, 218), the Secretary of the Treasury was authorized to borrow not exceeding \$400,000,000.

— Bayley, ch. 4, on the Act of June 30, 1864.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Five-Twenties of June 1864** are the second major "Five-Twenty" issue — a follow-on to the famous Five-Twenties of 1862 (ID 20101) that had been Jay Cooke's marketing triumph. Authorized under the **Act of June 30, 1864**, they carried the same structural features as their 1862 predecessor: a 6% coupon paid semi-annually in coin, callable after 5 years, final maturity 20 years.

The context, however, was entirely different. By mid-1864, the Civil War had entered its climactic phase. The Overland Campaign (Wilderness, Spotsylvania, Cold Harbor) had inflicted staggering Union casualties. Petersburg was under siege. Atlanta had not yet fallen. The 1862 Five-Twenties had exhausted their \$500 million authorization. New financing was urgent.

Chase, his reputation battered by the Ten-Forties episode, had resigned on June 30, 1864 — the same day the authorizing act was signed. His successor, **William Pitt Fessenden**, inherited the Treasury in a moment of crisis. The war was extraordinarily expensive, Lincoln's re-election was not yet certain, and investor fatigue with coupon-dilution had been real with the 5% Ten-Forties. Fessenden's response was to revert to the familiar 6% coupon for the new Five-Twenties, making them a direct substitute for the earlier, well-loved 1862 issue.

Peak outstanding was **\$125.6 million** — substantial, but only a fraction of the 1862 issue's \$515 million. Unlike the 1862 Five-Twenties, this bond has **no recorded secondary-market price series** in the Hall-Payne-Sargent database, which is consistent with its shorter lifetime and faster redemption (final database observation in January 1876, barely eleven years after issuance).


## Fessenden's Hand

**William Pitt Fessenden** took over Treasury on July 5, 1864. A Maine Republican with extensive Senate experience (he had chaired the Senate Finance Committee), Fessenden was less visionary than Chase but more skilled at working with Congress and the banking community.

His first major financing decision was the handling of the Act of June 30, 1864. Rather than risk another Ten-Forties–style coupon experiment, Fessenden opted for a **conservative design**: 6% coupon, 5-20 term, call feature — in structure identical to the 1862 issue. His calculation was straightforward: with the war in a decisive phase, maximum fundraising speed was worth more than a marginal reduction in interest cost.

The strategy largely worked. The new Five-Twenties found a ready market. Many investors treated them as interchangeable with the 1862 issue (they are sometimes grouped together in contemporary accounts). By spring 1865, outstanding balance was climbing rapidly.

### The Sister Issues of 1864–1865

The Act of June 30, 1864 authorized multiple instruments. Alongside this Five-Twenty, the Act supported:
- **Seven-Thirties of 1864 and 1865** (Bond ID 20111) — 3-year 7.3% notes, a reprise of the 1861 Seven-Thirties, designed as a convertible instrument.
- Additional Treasury notes and certificates for short-term liquidity.

As with the Act of March 3, 1863, the design was a **multi-instrument package** — Fessenden's approach was to give the Treasury multiple levers simultaneously, allowing allocation to the instrument finding the best market reception at each moment.

## Quantity Outstanding

Outstanding balance rose rapidly from March 1865 through 1867, reaching its peak near \$125 million. The post-1869 decline reflects two forces: (i) the Public Credit Act made refunding at lower coupons economically attractive, and (ii) the Funding Act of 1870 provided explicit statutory authority to redeem 6% Civil War debt and replace it with new 4%, 4.5%, or 5% long-term bonds. By 1876, this Five-Twenty had largely been refinanced.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20110, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Five-Twenties of June 1864 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20110, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.1f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.2f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Price Data

The Hall-Payne-Sargent database does not record a secondary-market price series for this specific issue. In contemporary market reports, the Five-Twenties of June 1864 were often **bundled with the 1862 Five-Twenties** in quoted indexes, since they had the same coupon, call structure, and maturity. Investors treated them as fungible. For the traded-price view of this family, readers should consult the chapter on Bond ID 20101 (the 1862 Five-Twenty).


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20110, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
    ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Five-Twenties of June 1864 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Five-Twenties of June 1864 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Legacy and Significance

### A Second Act for the Five-Twenty Design

The Five-Twenties of June 1864 confirmed what the 1862 issue had suggested: the **5-year call / 20-year maturity / 6% coin coupon** structure was the Union's most robust long-bond design. The market understood it, priced it consistently, and accepted new tranches without demanding punitive discounts.

### Fessenden's Pragmatism

By reverting to 6% rather than pushing further on the 5% front, Fessenden demonstrated a principle that would guide later Treasury secretaries: **in emergencies, reliability beats optimization**. A slightly higher coupon was a small price to pay for assured placement during the war's final phase.

### Quiet Refunding

Unlike the 1862 Five-Twenties — whose redemption became a national political question — the 1864 Five-Twenties were retired quietly through the refunding operations of the early 1870s. They played no role in the gold-vs-greenback debate, attracted no Cooke-style marketing attention, and left behind no special political legacy. They were, simply, workmanlike wartime finance.
