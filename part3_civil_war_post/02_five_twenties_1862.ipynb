## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The success of this loan was remarkable. Secretary Chase having used every exertion to provide for its general distribution among the people. In his annual report to Congress, dated December 10, 1863, he said: "After ascertaining, by inquiry, that they could not be disposed of to capitalists, in amounts sufficient for prompt payment of the army and navy, and for the satisfaction of the just claims of public creditors generally, without serious loss, the Secretary determined to employ a general agent, under adequate bonds, and confide the whole work of distribution… to him, and to sub-agents designated by him, and responsible immediately to him."

— Bayley, ch. 4, on the Five-Twenties of 1862, p. 65.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Five-Twenties of 1862** are the iconic bond of the Civil War — and arguably the single most consequential sovereign-debt instrument in 19th-century American history. Authorized by the **Legal Tender Act of February 25, 1862** (the same statute that created the greenback), they carried a 6% coupon, were callable after five years, and would mature in twenty. Their \$500 million authorization dwarfed anything the U.S. Treasury had ever attempted.

Structurally, the bond was conventional enough: semi-annual coupons, coin-payable. But politically, economically, and socially, the Five-Twenties were revolutionary. They became the vehicle through which **Jay Cooke**, a Philadelphia banker of extraordinary marketing gifts, transformed the idea of federal debt from an elite instrument held by banks and wealthy merchants into a **democratic asset** owned by millions of ordinary Americans — Iowa farmers, Pennsylvania coal miners, New York clerks, Midwestern widows.

They also inaugurated a **debate over monetary sovereignty** that would not be resolved until the **Public Credit Act of 1869**. Because the principal-payment terms were deliberately ambiguous, a decade-long political fight erupted over whether the Treasury should redeem the bonds in gold coin or in depreciated greenbacks. That debate shaped the politics of the 1868 presidential election, contributed to the rise of the Greenback Party, and helped define the partisan cleavage between "hard money" Republicans and "soft money" Democrats that dominated American monetary politics until 1896.

In the Hall-Payne-Sargent database, the Five-Twenties of 1862 appear in **154 monthly price observations** (September 1862 – June 1875), with an average price of **\$110.56**, ranging from **\$94.00** (February 1863, early-war low) to **\$123.88** (July 1869, immediately after the Public Credit Act). Peak outstanding was **\$514.8 million** (September 1868); final retirement was complete by 1877.


## A Half-Billion-Dollar Gamble

By early 1862, the Union's fiscal position had moved from serious to desperate. The war consumed roughly **\$2 million per day**. Federal revenues were still well under \$1 million per day. Foreign credit was effectively closed: **Napoleon III** had explicitly prohibited the sale of U.S. bonds on the Paris Bourse, and City of London underwriters — many of whom held Southern cotton exposure — showed little appetite for Northern debt.

Into this void, Congress passed the **Legal Tender Act of February 25, 1862**. The Act is famous today for creating the greenback, but it also did something equally consequential: it authorized the Five-Twenties for up to **\$500 million**. For perspective, the entire federal budget in 1860 had been \$63 million. The government was asking the public to lend it **eight times** a single pre-war year's expenditures.

The Treasury would finance the war through a deliberate three-legged stool:
- **Taxes** (notably the new 3% income tax)
- **Greenbacks** (non-interest-bearing fiat paper)
- **Bonds** (interest-bearing, long-dated — above all, the Five-Twenties)

## Jay Cooke — America's First Bond Salesman

No biography of the Five-Twenties can ignore **Jay Cooke**, a Philadelphia banker who turned an anemic bond float into a national cause. Cooke's achievements went far beyond mere sales:

- **2,500 salesmen** deployed across the Northern states
- **Newspaper alliances** (through his brother Henry Cooke's media connections) ensured patriotic messaging
- **Free coffee and donuts** were distributed at factory-shift changes to reach working-class subscribers
- When Five-Twenties traded as low as \$94.5 in March 1863, **Cooke personally bought every bond offered below par** — stabilizing the price and signaling his own financial commitment

Sales surged to **\$1 million per day**. Contemporary observers noted that "the great difficulty was to obtain enough bonds from the Treasury Department, which in turn was unable to get deliveries from the printers and engravers."

### The Gold Promise

In a widely reprinted letter, Cooke assured subscribers that **"Congress has provided that the Bonds shall be PAID in Gold when due."** This was an *interpretation*, not a statute: Congress had specified that *coupons* be paid in coin, but had pointedly left the redemption medium of the *principal* ambiguous. Cooke's public insistence that principal would also be repaid in gold effectively pre-committed the government to a harder line than Congress had legislated.

## Bond Quantity Over Time

The Five-Twenties' outstanding balance tells the story of the war economy in three phases:

1. **Issuance (late 1862 – early 1864):** Rapid climb from zero to the \$500M authorization ceiling, driven by Cooke's sales network and the National Banking Act's captive demand.
2. **Plateau (1864–1868):** Outstanding balance hovered near \$515M (the authorization plus a small premium) as the war concluded and the Treasury began contemplating how to refinance.
3. **Retirement (1869–1877):** Starting with the Public Credit Act, the Treasury used surplus revenues and refunding issues (4%, 4.5%, and 5% bonds under the 1870 Funding Act) to retire the Five-Twenties. Final redemption was complete by 1877.


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
for stype, color, ls in [('Public Holdings', '#1f77b4', '-'),
                          ('Total Outstanding', '#ff7f0e', '--')]:
    try:
        s = BondQuant_T.loc[(20101, stype)]
        s = s[s.notna()].sort_index()
        if len(s) > 0:
            ax.plot(s.index, s.values / 1e6, linewidth=2, linestyle=ls, color=color,
                    label=stype, marker='.', markersize=3)
    except KeyError:
        pass

ax.set_title('Five-Twenties of 1862 — Outstanding Quantity', fontsize=14, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Amount ($ millions)')
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        s = BondQuant_T.loc[(20101, stype)]
        s = pd.to_numeric(s, errors='coerce').dropna()
        if len(s) > 0:
            print(f"{stype}: max ${s.max()/1e6:.1f}M on {s.idxmax().strftime('%Y-%m-%d')}, {len(s)} obs, final ${s.iloc[-1]/1e6:.2f}M on {s.index[-1].strftime('%Y-%m-%d')}")
    except KeyError:
        pass
```

## Secondary Market Price

The price trajectory of the Five-Twenties is dominated by two forces that pushed in the same direction: (i) the depreciation of greenbacks against gold, which raised the greenback-denominated value of coin-payable bonds, and (ii) the gradual resolution of the gold-redemption question in favor of coin.

The low (~\$94, Feb 1863) occurred during the worst stretch of war finance — right before Cooke's network ramped up. The high (~\$124, July 1869) came three months after the Public Credit Act removed the last remaining doubt about redemption medium.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20101, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
    ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Five-Twenties of 1862 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Five-Twenties of 1862 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Price and Quantity Together

Overlaying the two series shows an important stylized fact about mid-19th-century Treasury debt: **price appreciation continued even as outstanding quantity plateaued**. Supply-side scarcity (via the \$500M ceiling) combined with demand-side expansion (captive bank demand + public appetite + speculative capital) produced a sustained premium through the 1868–1872 window.


```python
fig, ax1 = plt.subplots(figsize=(14, 6))

ax1.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c')
ax1.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6')
for stype in ['Total Outstanding', 'Public Holdings']:
    try:
        q = BondQuant_T.loc[(20101, stype)]
        q = q[q.notna()].sort_index()
        if len(q) > 0:
            ax1.fill_between(q.index, q.values / 1e6, alpha=0.2, color='#3498db')
            ax1.plot(q.index, q.values / 1e6, color='#3498db', alpha=0.8,
                     linewidth=1.5, label=f'Quantity ({stype})')
            break
    except KeyError:
        pass

ax1.set_xlabel('Date')
ax1.set_ylabel('Quantity Outstanding ($ millions)', color='#3498db')
ax1.tick_params(axis='y', labelcolor='#3498db')

ax2 = ax1.twinx()
try:
    p = BondPrice_T.loc[(20101, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
    if len(p) > 0:
        ax2.plot(p.index, p.values, color='#e74c3c', marker='.', markersize=3,
                 linewidth=2, label='Average Price')
        ax2.axhline(100, color='gray', alpha=0.3, linestyle='--')
        ax2.set_ylabel('Price ($)', color='#e74c3c')
        ax2.tick_params(axis='y', labelcolor='#e74c3c')
except KeyError:
    pass

ax1.set_title('Five-Twenties of 1862 — Price and Quantity Together', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right', fontsize=9, framealpha=0.9)
plt.tight_layout()
plt.show()
```

## Volatility & Drawdown

The rolling 12-month volatility reveals two clear spikes: (1) the early-war nadir of early 1863 and (2) the Panic of 1873, when Jay Cooke's own firm failed and triggered the Long Depression. Between those episodes, price behavior was relatively stable, consistent with a bond that had become the benchmark Union Treasury issue.


```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20101, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## The Gold-vs-Greenback Debate

No Civil War bond had greater political consequences than the Five-Twenties. Because Congress left the **principal redemption medium ambiguous**, a decade-long fight broke out:

- **Gold camp:** Northern financiers (led by Cooke), Republicans, and sound-money advocates argued that gold redemption was implicit in the original contract and necessary for preserving federal credit.
- **Greenback camp:** Western debtor farmers, Democrats, and the populist press (the *Cincinnati Daily Enquirer* under Washington McLean was particularly forceful) argued that redemption in greenbacks was legal, fair to wartime taxpayers, and would ease the chronic Western scarcity of specie.

The debate shaped the **Pendleton Plan** (named for Ohio Democrat George Pendleton), a central issue in the 1868 presidential election. Ulysses S. Grant's victory and his immediate signing of the **Public Credit Act of March 18, 1869** resolved the question in favor of gold — a decision that carried enormous distributional consequences, transferring wealth from greenback-holding debtors to bondholding creditors.

## Legacy

The Five-Twenties of 1862 were the prototype of **mass public finance** in America. The techniques Cooke pioneered — nationwide sales networks, celebrity endorsements, patriotic messaging, small denominations, installment purchasing — became the template for every subsequent mass bond campaign: Liberty Loans (1917–1919), War Bonds (1941–1945), and modern Savings Bonds.

The bond also established a crucial political precedent: **the federal government would honor its debts in the hardest currency available**. That commitment, purchased at considerable political and distributional cost, underwrote American sovereign credibility in international capital markets for over a century.

Ironically, the man who engineered the triumph did not enjoy its fruits. **Jay Cooke & Co. collapsed on September 18, 1873**, triggering the Panic of 1873 and the Long Depression. Cooke had over-extended financing the Northern Pacific Railroad. The Five-Twenties he had sold were redeemed in gold, as he had promised; their broker died in modest circumstances, his reputation eventually restored by historians.
