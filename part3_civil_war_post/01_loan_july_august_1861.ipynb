## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The extra session of Congress which began July 4, 1861, was held under very different circumstances from those which had accompanied any preceding session. Eleven states had attempted to secede from the Union, and were in actual rebellion. Seventy-five thousand volunteers had been called into the field to suppress the unlawful combination, and Congress met in a military camp.

— Bayley, ch. 4, p. 23, opening of the Civil War financing chapter.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Loan of July and August 1861** was among the earliest bonds floated after the outbreak of the Civil War — a direct response to the fiscal catastrophe that confronted the Union within weeks of Fort Sumter. Authorized by Congress under the **Act of July 17, 1861**, it belonged to a legislative package of unprecedented scope: together with the Seven-Thirties of 1861 and the Old Demand Notes, the July 17 Act authorized up to **\$250 million** in borrowing — roughly four times the size of the entire federal budget in 1860.

Secretary of the Treasury **Salmon P. Chase**, an Ohio Republican with limited prior experience in finance, found himself managing a wartime Treasury with an empty till. The Buchanan administration had left Chase almost nothing: a depleted cash reserve, demoralized credit, and a Southern banking system now in open rebellion. Chase's first bond floats, including the Loan of July and August 1861, were designed as **long-dated, coin-payable instruments** — reassuringly traditional in structure, offering 6% interest payable twice a year in gold.

The bond carried a **20-year term**, redeemable at the government's option after 1881. This structure reflected Chase's hope that a relatively quick war and a prompt return to normal finance would allow the debt to be refinanced at lower rates once victory was achieved. That hope would prove optimistic: the war would last four years, swell federal spending to an annual peak near \$1.3 billion, and end with over \$2.7 billion in federal debt outstanding — most of it owed to ordinary Americans who had never before held a government security.

On the secondary market, the Loan of July and August 1861 traded quite close to par in its brief recorded window (February 1862 through December 1862). The database records **11 monthly observations** for this bond, with an average of **\$99.77** and a range of **\$90.00–\$104.56**. Its peak outstanding balance was **\$189.3 million**, and a substantial portion remained in public hands for decades before being fully retired in the early 1890s.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> It authorized the Secretary of the Treasury to borrow, on the credit of the United States, within twelve months from the passage of the act, a sum not exceeding \$250,000,000, for which he was authorized to issue coupon or registered bonds, or treasury notes, in such proportions of each as he might deem advisable, the bonds to bear interest not exceeding 7 per cent. per annum, payable semi-annually, and redeemable, at the pleasure of the United States, after twenty years. The treasury notes to be of any denomination fixed by the Secretary, not less than \$50, payable three years after date, with interest at the rate of 7 3/10 per cent. per annum, payable semi-annually.

— *Bayley, ch. 4 (Loan of July and August 1861), p. 23 — Act of July 17, 1861 (12 Statutes, 259).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> A bill to meet the expenses of the war, by obtaining a loan of \$250,000,000, was introduced in the House July 9, 1861, and considered in committee of the whole the next day. Debate on the bill was limited to one hour, and this hour was entirely occupied by one member. At the conclusion of his remarks the bill was passed by yeas 150, nays 5.

— *Bayley, ch. 4 — the extraordinary single-hour debate and 150-5 House vote that authorized the largest federal borrowing in American history up to that date.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Under these acts bonds and treasury notes were issued as follows: Loan of July and August, 1861, 6 per cent. bonds, \$189,321,350; Old demand notes, no interest, 60,030,000; Seven-Thirties of 1861, interest at 7 3/10 per cent., 140,094,750. Total: \$389,446,100. When it is considered that only a year before a loan of \$21,000,000 had been asked for, and only about one-third that sum obtained, it is evident that the danger which menaced the Union had aroused the people and induced them to pour out their means freely in aid of the government.

— *Bayley, ch. 4 — the issuance breakdown across the three instruments authorized by the Act of July 17, 1861, totaling \$389M.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

### Political and fiscal context

Bayley also frequently records the political backdrop — the wartime urgency, the legislative debate, the economic conditions — that shaped the placement:

> The disasters of the early part of the war were really a gain instead of a loss, as they convinced the North of the necessity of freely contributing money and men to save the imperiled nation.

— *Bayley, ch. 4 — Bayley's editorial reflection on how military setbacks paradoxically aided wartime financing.*

Read the modern paragraphs that follow with these four Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## The Fiscal Crisis of 1861

When Lincoln took the oath of office on **March 4, 1861**, the Treasury was, in plain language, broke. The federal government had been running structural deficits since the Panic of 1857. Foreign credit was virtually closed: European investors — especially in London and Paris — were skeptical of a country whose territorial integrity was suddenly in question. Domestic credit was little better: the major New York banks had been burned in earlier attempts to finance the outgoing Buchanan administration at punitive rates (some short-term notes had paid coupons as high as 12%).

### The Act of July 17, 1861

Less than three months after Fort Sumter, Congress passed the landmark **Act of July 17, 1861**, one of the most consequential fiscal statutes in American history. It authorized Chase to borrow up to **\$250 million** in a combination of three instruments:

| Instrument | Coupon | Term | Role |
|------------|--------|------|------|
| **20-year bonds** (this issue) | 6.0% | 20 years | Long-dated, coin-payable — this bond |
| **7.30% three-year Treasury notes** ("Seven-Thirties") | 7.3% | 3 years | Mid-term, aimed at banks and wealthier investors |
| **Old Demand Notes** | 0.0% | on demand | Early paper currency, non-interest-bearing |

The bonds were marketed domestically. Chase initially envisioned that New York banks would absorb large tranches of the debt and redistribute them to the public. In practice, banks were willing to underwrite only at substantial discounts, and the early tranches moved slowly.

### Who Bought?

The original subscribers were overwhelmingly **banks and wealthy Eastern financiers**. Unlike the later Five-Twenties (which Jay Cooke would market as a mass-democratic bond), the Loan of July and August 1861 was still being sold through traditional channels: private banking houses, bond syndicates, and individual capitalists. The absence of a Jay Cooke–style marketing apparatus at this early stage of the war meant that sales were brisk but uneven, and fully placing the \$189 million of eventual outstanding balance took several months.

## The Issuance Sequence under the Act of July 17, 1861

The July 17 Act was not a single bond issue but a **fiscal toolkit**. Chase used different instruments for different audiences:

1. **Seven-Thirties of 1861** (Bond ID 20100) — issued first. The 7.30% coupon (selected because it equals 2 cents per \$100 per day — easy mental arithmetic at a time when newspapers printed "yield conversion tables" for laymen) and 3-year term made these notes attractive to banks as a liquidity instrument.
2. **The Loan of July and August 1861** (this bond) — the long-dated, coin-payable companion issue. Slower to place but designed to lock in 20-year financing.
3. **Old Demand Notes** (Bond ID 30002) — the Treasury's first experiment with non-interest-bearing paper. Initially receivable for public dues, they anticipated the greenback innovation of 1862.

This triad allowed Chase to calibrate issuance to demand: when banks resisted the 20-year bonds, he could lean on the short-term notes; when bank deposits were short, he could issue demand notes.

## Secondary Market Price

The database records a short but illuminating **11-month price series** for this bond, running from February 1862 to December 1862. The range (\$90.00–\$104.56) is tight, and the average (\$99.77) is remarkably close to par for a bond that traded through the first year of the Civil War.

Three points about this price series are worth emphasizing:

1. **The February 1862 low (\$90.00)** coincides with an unusually fraught moment: Northern banks had just suspended specie payments, the Legal Tender Act was still being debated in Congress, and Union military fortunes in the Western theater had not yet shifted. Investors rationally demanded a small discount for holding a 20-year, coin-payable bond whose redemption medium could, in theory, be altered by Congress.

2. **The June 1862 high (\$104.56)** reflects a combination of factors: the Legal Tender Act had now passed (providing a backstop liquidity instrument), and the greenback's depreciation had begun to make gold-payable bonds more valuable in paper-dollar terms. The resulting premium is a classic example of how a coin-payable instrument appreciates against a fiat alternative.

3. **The brevity of the series** reflects a documentary limitation of the database, not necessarily market reality: later observations were apparently not collected for this specific issue, perhaps because trading volume migrated to the larger Five-Twenty and Seven-Thirty issues. After 1862, most secondary-market activity for Union long-dated debt seems to have centered on the Five-Twenties of 1862 (ID 20101), which dominated the government-bond market for the remainder of the war.

The bond's price reaches its 1862 high as outstanding quantity is still climbing — a sign that demand was absorbing issuance without forcing meaningful discounts. ## Volatility and Drawdown

With only 11 observations, formal volatility estimation is fragile. The rolling 12-month window cannot be computed here. Nevertheless, the total trading range (roughly 14 points, from \$90 to \$104) is consistent with a wartime long-bond market in which confidence in the government's solvency was still being tested.


```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20099, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

The Loan of July and August 1861 occupies a unique place in American fiscal history for several reasons:

### 1. First Civil War Long-Bond

It was the **first 20-year coin-payable bond issued for the Union war effort**. Its successful (if gradual) absorption proved that a sufficient domestic market existed for long-dated sovereign debt even under extreme wartime uncertainty. This provided the template — though not yet the marketing machinery — that Jay Cooke would scale up dramatically with the Five-Twenties in 1863.

### 2. Bridge Between Old and New Treasury Finance

Structurally, the bond looks backward: a conventional 20-year instrument, sold through banking syndicates, with coin-payable coupons. In its role within the July 17 Act, however, it looks forward: it was one leg of a three-part financing stool that included Treasury notes and demand notes, anticipating the diversified wartime issuance strategy that would characterize the rest of the war.

### 3. Part of the Pre-Cooke Old Guard

Unlike the mass-marketed Five-Twenties, this bond was sold primarily to banks and wealthy investors through traditional underwriting. It therefore represents the **last major wartime bond issued through the old financial architecture**. Within eighteen months, Jay Cooke's national sales network would have redefined what government-bond marketing looked like.

### 4. A Surviving Piece of Funded Debt

The bond's quiet presence on the Treasury's books well into the 1880s and 1890s shows how the **Funding Act of 1870** was used to refinance old Civil War debt into lower-coupon instruments. Many of the holders who continued to own these bonds chose not to participate in the refunding operations — either because they valued the 6% coupon, were unaware of the refunding offers, or simply lost track of their certificates. Federal debt of the 19th century often had long, quiet tails of this kind.
