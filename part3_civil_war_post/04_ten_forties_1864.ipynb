## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> By the act of March 3, 1864 (13 Statutes, 13), an additional \$200,000,000 was authorized, in bonds redeemable after ten and payable in forty years, bearing not exceeding 6 per cent. interest, payable semi-annually, in coin.

— Bayley, ch. 4, on the Act of March 3, 1864.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Ten-Forties of 1864** are an unusual — and in some ways tragic — instrument. They were Secretary **Salmon P. Chase's** most ambitious attempt to lower the Treasury's borrowing costs, and in a narrow technical sense they accomplished that goal: they carried a **5% coupon**, well below the 6% rate that had dominated Union long-term finance since 1861. But the lower coupon proved, at least initially, **too aggressive for the market**. When the bonds were offered at par in the spring of 1864, sales were sluggish. Chase was forced to rely on extraordinary measures — including the personal intervention of Jay Cooke — and, ultimately, to resign.

His successor, **William Pitt Fessenden**, took over the Treasury on July 5, 1864. Fessenden, the former chairman of the Senate Finance Committee, had a deeper relationship with Congress than Chase and proved more flexible in managing wartime debt. Under Fessenden, the Ten-Forties were successfully placed — though at roughly \$195 million they came in well below the original hopes. His successor, **Hugh McCulloch**, would complete the placement and supervise the secondary-market stabilization of the 1865–1866 period.

The bond's name — "Ten-Forties" — refers to its unusual call/maturity structure: **callable after 10 years, final maturity at 40 years**. This was an attempt to offer the Treasury long-run option value while still giving investors a clear horizon. The extended maturity, unusual for the time, reflected Chase's belief that post-war refinancing opportunities would make the 40-year endpoint largely theoretical.

In the Hall-Payne-Sargent database, the Ten-Forties trade through **174 monthly price observations** spanning August 1864 to April 1879. Mean price: **\$107.68**. Range: **\$91.00** (November 1865, early postwar volatility) to **\$119.39** (January 1876, after substantial deflation and refunding announcements). Peak outstanding: **\$194.6 million**.


## The Coupon Controversy

The most interesting feature of the Ten-Forties is how **nearly they failed**. Chase's hypothesis in proposing a 5% coupon was that the market had, by early 1864, accepted the government's credit and would accept a lower yield on long-dated securities. The hypothesis was half right.

### Initial Resistance

When the Ten-Forties were first offered in the spring of 1864, investors **balked at the 5% coupon**. They had become used to 6% and, in the case of the Seven-Thirties, 7.3%. A 5% coupon felt like a concession just as the war was entering its most expensive phase (the Overland Campaign, the siege of Petersburg, Sherman's Atlanta campaign were all consuming cash at unprecedented rates).

### Cooke Declines the Call

Jay Cooke, asked to handle brokerage, initially **declined to aggressively market the Ten-Forties at 5%** — a reversal from his Five-Twenty evangelism. Cooke privately concluded the coupon was too low and publicly marketing it would damage his credibility with subscribers.

### Chase's Resignation

A combination of rising political friction with Lincoln (who had refused to accept Chase's earlier resignation threats as leverage), a tightening financial market, and the Ten-Forties' weak reception led Chase to submit his resignation on **June 30, 1864**. This time, Lincoln accepted. Fessenden took over on July 5.

### Fessenden's Rescue

Under Fessenden, the Ten-Forties were still offered at 5%, but placement was coupled with other instruments (notably the Five-Twenties of June 1864 at 6% and the Seven-Thirties of 1864 at 7.3%). The combined offering allowed investors to blend coupon risk across different issues. By year-end 1864, the Ten-Forties had found their market — helped significantly by the military momentum of Atlanta's fall (September 2, 1864) and Lincoln's re-election (November 8, 1864).

## Quantity Outstanding

Outstanding balance rose steadily from issuance through the late 1860s, peaking near \$195 million, and then remained stable until the Funding Act of 1870 began the refunding process. ## Secondary Market Price

The Ten-Forties' price history is a revealing window into the post-war bond market:

1. **1864–1865 Low (\$91):** Initial skepticism about the 5% coupon, confounded by general market turbulence at war's end.
2. **1866–1868 Recovery:** Prices climbed toward \$110 as the war ended and sovereign credibility improved.
3. **1869 Public Credit Act:** Prices spiked further as the gold-redemption question was resolved favorably.
4. **1876 Peak (\$119):** Post-deflation environment with 5% coupon increasingly attractive relative to falling market rates; refunding rumors also supported prices.
5. **1873–1879 Volatility:** The Panic of 1873 and Long Depression produced intermittent selling pressure, but the bonds never retraced to their pre-1869 lows.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20108, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1861-04-12'), pd.Timestamp('1865-05-09'), alpha=0.15, color='#e74c3c', label='Civil War')
    ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Ten-Forties of 1864 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Ten-Forties of 1864 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20108, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### Lower-Coupon Precedent

Despite their rocky debut, the Ten-Forties eventually established a crucial precedent: **the U.S. Treasury could issue long-dated debt at coupons below 6%**. This opened the door to the 4.5% and 4% refunding bonds of the 1870s — instruments that would substantially reduce the nation's interest burden as Civil War debt was refinanced.

### The 40-Year Maturity

A 40-year bond was unusual in 1864 and remained unusual for decades after. Only the refunding bonds of 1870 (notably the 4% Loan of 1907) would carry similar ultra-long maturities. The Ten-Forties thus sit in an interesting position within the term structure of 19th-century U.S. sovereign debt.

### End of the Chase Era

Symbolically, the Ten-Forties mark the end of Chase's tenure and the transition to a more professionalized Treasury under Fessenden and McCulloch. Chase's legacy as Treasury Secretary is complex: he presided over the greenback, the National Banking Act, the Five-Twenty campaign, and the initial framework for Civil War taxation. He also nearly failed with the Ten-Forties. His departure from Treasury (and elevation to the Chief Justiceship in December 1864) closed a distinct chapter in American fiscal history.
