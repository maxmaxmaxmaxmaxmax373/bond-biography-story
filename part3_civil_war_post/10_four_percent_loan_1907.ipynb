## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> The act of July 14, 1870 (16 Statutes, 272), authorized the issue of bonds, in three classes — fifteen years at four and a half per cent., ten years at five per cent., and thirty years at four per cent.

— Bayley, ch. 4, on the Funded Loans of 1881, 1891, and 1907.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Four Percent Loan of 1907** is the final and largest instrument of the **Funding Act of July 14, 1870** refunding program. With its **30-year term** and 4.0% quarterly coupon, it was explicitly designed to lock in low-cost financing for a generation. Its peak outstanding — **\$739.4 million** — made it the single largest bond issue of the late 19th century, surpassing even the Civil War–era 1862 Five-Twenties (\$515M) and the Consols of 1865 (\$333M).

The 4% Loan served as the anchor of the Funding Act's rate-compression strategy. The Act had created three refunding instruments: the 5% Loan of 1881 (intermediate), the 4.5% Loan of 1891 (medium-long), and this 4% Loan of 1907 (long). By issuing the 4% Loan at par with a 30-year redemption horizon, Treasury Secretary **John Sherman** effectively **extended the maturity structure of the federal debt by a full decade** while simultaneously cutting the coupon.

The bond's lifetime (1878–1907+) spans an extraordinary arc of American history: the resumption of specie payments (1879), the Garfield assassination (1881), the rise of Grover Cleveland, the Sherman Silver Purchase Act (1890), the Panic of 1893, William Jennings Bryan's "Cross of Gold" speech (1896), the Gold Standard Act (1900), the Spanish-American War (1898), and finally its own redemption in 1907 as the Panic of 1907 was brewing.

In the Hall-Payne-Sargent database, the 4% Loan appears in an astonishing **357 monthly price observations** — one of the longest price series in the database. Mean price: **\$114.11**. Range: **\$98.88** (March 1879, early months before specie resumption) to **\$129.52** (May 1887, the premium peak during the long deflation of the 1880s).


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> Also a sum or sums not exceeding in the aggregate one thousand million dollars of like bonds, the same in all respects, but payable at the pleasure of the United States after thirty years from the date of their issue, and bearing interest at the rate of four per cent. per annum; all of which said several classes of bonds and the interest thereon shall be exempt from the payment of all taxes or duties of the United States, as well as from taxation in any form by or under state, municipal, or local authority.

— *Bayley, ch. 4 (Funded Loans of 1881, 1891, and 1907) — quoting the 4% provision of the Funding Act of July 14, 1870 (16 Statutes, 272).*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> The United States of America are indebted to ___, or assigns, in the sum of ___. This bond is issued in accordance with the provisions of an act of Congress entitled "An act to authorize the refunding of the national debt, approved July 14, 1870", and an act amendatory thereof, approved January 20, 1871, and is redeemable at the pleasure of the United States, after the first day of July, A.D. 1907, in coin of the standard value of the United States on said July 14, 1870, with interest in such coin from the day of the date hereof at the rate of 4 per centum per annum, payable quarterly on the first day of October, January, April, and July of each year. The principal and interest are exempt from the payment of taxes or duties of the United States, as well as from taxation in any form by or under state, municipal, or local authority.

— *Bayley, ch. 4 — verbatim text of the Four Per Cent. Consol bond (dated July 1, 1877), with its quarterly-coupon and tax-exempt provisions.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Four per cent. loan of 1907: \$739,480,800. … In offering the new loan citizens and subjects of other governments should receive the strongest assurance that the interest and principal are to be paid in coin, according to the terms of the bonds issued, without any deduction or abatement whatsoever.

— *Bayley, ch. 4 — the cumulative issuance figure (\$739.5M) and Boutwell's signature commitment-to-coin language directed at international subscribers.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

### Political and fiscal context

Bayley also frequently records the political backdrop — the wartime urgency, the legislative debate, the economic conditions — that shaped the placement:

> Mr. Boutwell recommended that, in order to avoid the necessity of employing agents for the negotiation of the proposed new loan, a liberal commission be allowed to subscribers, and that those first subscribing be permitted to select the class of bonds in which their subscriptions should be made; and further, that the national banks be required to substitute bonds of the proposed new loan for such as were then deposited in the treasury as security for the redemption of their circulating notes.

— *Bayley, ch. 4 — Boutwell's distribution strategy: subscriber-class selection and forced national-bank reserve substitution.*

Read the modern paragraphs that follow with these four Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## Sherman's 30-Year Gamble

When Secretary **John Sherman** first offered the 4% Loan in March 1878, the decision was not universally hailed. A 4.0% coupon was lower than anything the U.S. had issued in half a century, and a 30-year term locked the Treasury into a very long-dated liability. Critics — including some in Sherman's own party — worried that the coupon would prove too low to attract adequate subscriptions, repeating the Ten-Forties experience of 1864.

### Why It Worked

Four factors ensured the 4% Loan's success:

1. **Specie resumption was imminent:** The Specie Payment Resumption Act of 1875 had set a hard deadline of January 1, 1879. As that date approached, gold-redemption risk premia collapsed. By late 1878, the market was pricing U.S. sovereign debt as *de facto* gold-standard paper.

2. **International investors returned:** With the gold standard restored, European capital flowed back into American markets. London and Paris houses found 4% U.S. sovereign debt highly attractive — safer than most European sovereigns and carrying a coupon above British Consols (then paying ~3%).

3. **Quarterly coupons and long duration:** Institutional investors (trust companies, insurance companies, savings banks) valued the combination of quarterly cash flow and 30-year horizon.

4. **Sherman's refunding syndicate:** Sherman worked closely with leading New York banking houses (notably **August Belmont & Co.** and **Drexel, Morgan & Co.**) to form a refunding syndicate that could absorb large tranches and redistribute them internationally.

### The Refunding Machine

The 4% Loan was the capstone of a sequence of refunding operations that, over 1877–1881, retired most of the outstanding Civil War 6% debt. The mechanics were straightforward: Treasury issued 4% Loans at par; holders of 6% Five-Twenties and Consols (Bond IDs 20101, 20110, 20113, 20114, etc.) exchanged their older bonds for new ones, typically at a cash premium that recognized the higher-coupon value being surrendered. Over four years, this machinery cut roughly **\$2 billion in federal debt** from 6% coupons to 4% coupons — an annual interest saving of approximately **\$40 million**.

## Quantity Outstanding

- **1878–1881:** Rapid growth from initial issuance to ~\$739M peak as Sherman's refunding machine worked through the Civil War 6% debt.

## Secondary Market Price

With 357 monthly observations, the 4% Loan has one of the richest price histories in the database. Several epochs stand out:

1. **March 1879 Trough (\$98.88):** The only sub-par observation in the entire series. Early months before specie resumption took hold.
2. **1880–1887 Premium Build:** Prices climbed from par to the \$129+ peak as the gold standard was restored, deflation set in, and international demand grew.
3. **1887 Peak (\$129.52):** The premium reflected both the quality of the 4% coupon in a deflationary environment and the scarcity of comparable long-duration gold-standard instruments.
4. **1890s Volatility:** Sherman Silver Purchase Act (1890), Panic of 1893, and Bryan's "Cross of Gold" populism all produced brief selling spikes as gold-standard risk re-emerged.
5. **1896–1900 Recovery:** McKinley's election (1896) and the Gold Standard Act (1900) restored confidence. Prices settled in the \$120s for the bond's final years.
6. **1907 Pull-to-Par:** As the redemption date approached, prices converged toward par.

## Legacy and Significance

### A Monument of Gilded Age Finance

The Four Percent Loan of 1907 stands as a monument of the Gilded Age federal debt-management strategy. No single instrument did more to **reduce the federal government's interest burden** after the Civil War. No single instrument provided a more stable benchmark for **long-duration gold-standard investment** during the deflationary decades of the late 19th century.

### Interest Savings

Cumulative savings from the 4% Loan, over its 30-year life, exceeded \$600 million relative to what the government would have paid on 6% Civil War debt. This was equivalent to roughly a full year of federal revenues in the mid-1880s.

### Benchmark for Future Borrowing

The successful placement of a 30-year, 4% coupon bond at par in 1878 established a template for later U.S. sovereign debt:
- The **Panama Canal Loans of 1906 and 1908** (IDs 20132, 20133) were issued at 2% — an even lower coupon, explicitly justified by the 4% Loan's precedent.
- The **Consols of 1930** (ID 20131) likewise invoked the 4% Loan as the model for long-term, low-coupon federal borrowing.
- The **Liberty Loans of 1917–1919** drew organizational lessons from the quarterly-coupon, large-denomination structure of the 4% Loan (though they adopted semi-annual coupons for political reasons).

### Retirement and Succession

The 4% Loan's final redemption in 1907 was quiet but symbolically important. At the moment of redemption, the Panic of 1907 was brewing — the crisis that would ultimately motivate the creation of the Federal Reserve System six years later. The bond thus bridges two eras of American central-banking history: it was issued in the immediate post-Civil War fiscal consolidation, and it was redeemed just as the modern Federal Reserve era was about to begin.
