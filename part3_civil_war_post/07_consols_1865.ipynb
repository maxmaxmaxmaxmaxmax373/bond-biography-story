## Overview

Rafael A. Bayley, writing in 1881 in his *National Loans of the United States*, framed this loan in the following terms:

> Under the authority conferred by these two acts the following 6 per cent. bonds have been issued: Five-twenties of 1865, \$203,327,250; Consols of 1865, \$332,998,950; Consols of 1867, \$379,618,000; Consols of 1868, \$42,539,350. Total, \$958,483,550.

— Bayley, ch. 4, on the Five-Twenties of 1865 and the Consols.

That contemporary framing is the place to begin. The modern paragraphs that follow extend it with the secondary-market data the Hall–Payne–Sargent Bond Database makes available, and with the broader 21st-century historiographical literature on the loan.

The **Consols of 1865** are the first and largest of three Consol series (1865, 1867, 1868) issued under the Act of March 3, 1865. The name "Consols" is borrowed from British fiscal tradition (*Consolidated Annuities*), where Consols had been perpetual securities paying a fixed coupon since 1751. The American Consols were not perpetual — they were 5-year-call/20-year-maturity bonds — but the name signaled the policy intent: **consolidation** of multiple outstanding Civil War debts into a smaller number of standardized instruments.

Issued beginning March 31, 1867, the Consols of 1865 carried the now-familiar 6% coin-payable coupon and 5-20 call/maturity structure. What distinguished them from their Five-Twenty siblings was primarily their **timing and role**: they were issued after the most urgent wartime financing was complete, during the intense refunding and reorganization of Civil War debt that characterized McCulloch's tenure.

Peak outstanding reached **\$332.5 million** — the largest of any single Civil War–era bond save the original 1862 Five-Twenties. The database records **141 monthly price observations** (February 1867 – November 1878), with mean price **\$112.80**, range **\$102.50–\$123.07**. Like their Five-Twenty siblings, the Consols traded at premium throughout their life, reflecting the durable value of the 6% coin coupon as market rates declined through the late 1860s and 1870s.


## Bayley's Account in Detail (1881)

Bayley's 1881 *National Loans of the United States* devotes substantive attention to this bond. Four passages from his account are reproduced below, with modern commentary woven in. They give the reader the primary-source spine of this biography — the statute text, the placement mechanics, and the post-issuance disposition as recorded by the U.S. Treasury Department officer who compiled the original.

### The authorizing statute

Bayley begins his account, as he does for every loan in his book, with the statutory text. For this bond he writes:

> Be it enacted, &c., That the act entitled "An act to provide ways and means to support the government", approved March third, eighteen hundred and sixty-five, shall be extended and construed to authorize the Secretary of the Treasury, at his discretion, to receive any treasury notes or other obligations issued under any act of Congress, whether bearing interest or not, in exchange for any description of bonds authorized by the act to which this is an amendment.

— *Bayley, ch. 4 — quoting the Act of April 12, 1866 (14 Statutes, 31), the supplementary statute under which the Consols were placed.*

That paragraph is the legal foundation of the bond. Every subsequent operational and financial detail in this biography follows from the statutory powers it conferred on the Treasury.

### How it was actually placed

Bayley then describes how the Treasury translated statutory authority into actual placement — the subscribers, the marketing, the prices realized, the political context that shaped market reception:

> Next in importance to the great questions involved in the restoration of the federal authority over the Southern states, and the re-establishment of civil government therein under the constitution, are the financial questions, embracing the currency, the public debt, and the revenue; all of which demand the early and careful attention of Congress. The fact that means have been raised, without foreign loans, to meet the expenses of a protracted and very costly war is evidence, not only of the great resources of the country, but of the wisdom of Congress in passing the necessary laws.

— *Bayley, ch. 4 — Treasury Secretary Hugh McCulloch, *Annual Report*, December 4, 1865 — opening the postwar refunding case.*

This operational detail is the kind of material that distinguishes Bayley's primary-source account from purely statistical sources. The price and quantity series in the Hall–Payne–Sargent database tell us *what* happened in the secondary market; Bayley tells us *how* it happened in the primary placement.

### Outstanding amounts and disposition

Bayley closes most of his loan accounts with concrete tabulated figures — total amount issued, distribution among holders, ultimate retirement schedule:

> Secretary McCulloch recommended: First. That Congress declare the compound-interest notes shall cease to be a legal tender from the day of their maturity. Second. That the Secretary be authorized, in his discretion, to sell bonds of the United States, bearing interest at a rate not exceeding 6 per cent., and redeemable and payable at such periods as may be conducive to the interests of the government, for the purpose of retiring not only compound-interest notes, but the United States notes.

— *Bayley, ch. 4 — McCulloch's specific December 1865 recommendations that produced the April 1866 act and the Consol issuance.*

These figures are the data that the modern Hall-Payne-Sargent database extends and corroborates. Where Bayley's tabulated total matches the database's monthly outstanding peak (within rounding), we have strong confidence in the bond's accounting; where they differ, the discrepancy itself carries information about how 19th-century federal debt was tracked.

Read the modern paragraphs that follow with these three Bayley passages in mind. The 21st-century historiography extends Bayley's account; it does not replace it.

## Why "Consols"?

The name tells a policy story. After the Civil War, the U.S. Treasury held an extraordinary variety of outstanding obligations: 6% bonds, 7.3% notes, 5% bonds, Compound Interest Notes, Temporary Loans, Certificates of Indebtedness, numerous state-level war debts assumed by the federal government, and more. Each instrument had its own coupon, maturity, call feature, and tax status. This complexity had been **necessary during wartime** (when issuance had to be opportunistic) but was **undesirable in peacetime** — it created operational burdens for the Treasury and opacity for investors.

### The Consolidation Idea

The Act of March 3, 1865 authorized Treasury to exchange outstanding Civil War obligations for new "consolidated" bonds with **standardized features**. Holders of various older instruments could present them to the Treasury and receive Consols in exchange. In this sense, the Consols functioned like a **debt-exchange mechanism** — a tool for simplifying the federal balance sheet.

### A Partial Consolidation

In practice, the Consols did not fully consolidate Civil War debt. The Five-Twenties of 1862 remained largely intact. The Seven-Thirties were converted into Five-Twenties rather than Consols. The Ten-Forties remained in circulation. What the Consols did accomplish was to **establish a standardized 6%-coin-payable benchmark instrument** that would become, through its sheer size and stability, the reference bond for subsequent refunding operations under the 1870 Funding Act.

## Quantity Outstanding

The Consols of 1865 grew rapidly from first issue in March 1867, reaching their peak near \$333 million by the early 1870s. This is a reflection not only of new issuance but of **exchanges**: holders of older Civil War debt could convert into Consols, adding to the outstanding stock without requiring new cash subscriptions. The post-1873 decline reflects accelerated refunding under the Funding Act.

## Secondary Market Price

The Consols traded from \$102.50 (July 1878, during the Long Depression) to a peak of \$123.07 (June 1875). Over 141 observations, mean price was \$112.80. Key inflection points:

- **March 1867 First Issue:** Bonds opened at a modest premium (~\$103), reflecting the already-established quality of the 6% coin structure.
- **1869 Public Credit Act Rally:** Prices jumped as gold redemption was mandated.
- **1875 Peak:** Post-Resumption Act environment combined with falling market rates drove prices to \$123.
- **1878 Decline:** Long Depression selling, combined with refunding expectations (holders anticipated redemption at par under the Funding Act), drove prices back down toward \$102.


```python
fig, ax = plt.subplots(figsize=(14, 7))

try:
    s = BondPrice_T.loc[(20114, 'Average')]
    s = pd.to_numeric(s, errors='coerce').dropna().sort_index()
except KeyError:
    s = pd.Series(dtype=float)

if len(s) > 0:
    ax.plot(s.index, s.values, marker='.', markersize=5, linewidth=2,
            color='#2c3e50', label='Average Price', zorder=5)
    ax.axhline(100, color='black', alpha=0.3, linestyle='--', linewidth=1.5, label='Par ($100)')

    ax.axvspan(pd.Timestamp('1865-05-10'), pd.Timestamp('1877-03-31'), alpha=0.15, color='#f39c12', label='Reconstruction')
    ax.axvspan(pd.Timestamp('1873-09-18'), pd.Timestamp('1879-03-01'), alpha=0.15, color='#9b59b6', label='Panic of 1873 / Long Depression')
    pmin_d, pmin_v = s.idxmin(), s.min()
    pmax_d, pmax_v = s.idxmax(), s.max()
    ax.annotate(f'LOW: ${pmin_v:.2f}\n({pmin_d.strftime("%b %Y")})',
                xy=(pmin_d, pmin_v), xytext=(30, -35),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#c0392b',
                arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))
    ax.annotate(f'HIGH: ${pmax_v:.2f}\n({pmax_d.strftime("%b %Y")})',
                xy=(pmax_d, pmax_v), xytext=(30, 20),
                textcoords='offset points', fontsize=10, fontweight='bold', color='#27ae60',
                arrowprops=dict(arrowstyle='->', color='#27ae60', lw=1.2))
    ax.set_title('Consols of 1865 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Price ($)')
    ax.legend(loc='lower left', fontsize=9, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    print(f"Price Statistics ({len(s)} obs):")
    print(f"  Mean ${s.mean():.2f} | Std ${s.std():.2f} | Range ${s.min():.2f}-${s.max():.2f}")
    print(f"  First obs: {s.index.min().strftime('%Y-%m-%d')} | Last obs: {s.index.max().strftime('%Y-%m-%d')}")
else:
    ax.text(0.5, 0.5, 'No secondary-market price series is recorded\nfor this bond in the Hall-Payne-Sargent database.',
            transform=ax.transAxes, ha='center', va='center', fontsize=14,
            bbox=dict(boxstyle='round,pad=0.5', fc='#fef9e7', ec='#f39c12'))
    ax.set_title('Consols of 1865 — Secondary Market Price', fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Volatility & Drawdown

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})

try:
    p = BondPrice_T.loc[(20114, 'Average')]
    p = pd.to_numeric(p, errors='coerce').dropna().sort_index()
except KeyError:
    p = pd.Series(dtype=float)

ax = axes[0]
if len(p) > 0:
    ax.plot(p.index, p.values, color='#2c3e50', linewidth=2, label='Price')
    ax.axhline(100, color='black', alpha=0.2, linestyle='--')
    running_max = p.expanding().max()
    ax.fill_between(p.index, p.values, running_max.values, alpha=0.3,
                    color='#e74c3c', label='Drawdown from Peak')
    ax.set_title('Price and Drawdown from Peak', fontsize=13, fontweight='bold')
    ax.set_ylabel('Price ($)'); ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, 'No price series available', transform=ax.transAxes, ha='center', fontsize=12)

ax2 = axes[1]
if len(p) >= 12:
    rolling_std = p.rolling(window=12).std().dropna()
    ax2.fill_between(rolling_std.index, rolling_std.values, alpha=0.4, color='#e74c3c')
    ax2.plot(rolling_std.index, rolling_std.values, color='#c0392b', linewidth=1.5)
ax2.set_title('12-Month Rolling Price Volatility (Std Dev)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Date'); ax2.set_ylabel('Std Dev ($)'); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

if len(p) > 0:
    dd = (p.min() - p.max()) / p.max() * 100
    print(f"\nMaximum drawdown from rolling peak: {dd:.1f}%")
    print(f"  Peak ${p.max():.2f} on {p.idxmax().strftime('%Y-%m-%d')}")
    print(f"  Trough ${p.min():.2f} on {p.idxmin().strftime('%Y-%m-%d')}")
```

## Legacy and Significance

### Policy Template for Debt Management

The Consols of 1865 demonstrated the power of **standardization** in sovereign debt management. By consolidating multiple disparate obligations into a single instrument class, the Treasury:
- Simplified its operational burden
- Improved market liquidity in a single benchmark security
- Created a reference yield curve that clarified the true cost of sovereign borrowing
- Positioned outstanding debt for subsequent refunding

These lessons would be applied repeatedly in later refundings and, more than a century later, in modern Treasury debt management through standardized maturities and reopenings of on-the-run issues.

### Size and Scale

At \$332.5 million peak outstanding, the Consols of 1865 were one of the largest single bond issues in 19th-century U.S. history. Only the 1862 Five-Twenties (at \$515 million) were larger. Together, these two instruments formed the backbone of the funded federal debt through the Reconstruction era.

### Gateway to the Funding Act

The Consols of 1865 were the primary target of the **Funding Act of July 14, 1870**, which authorized their retirement and replacement with lower-coupon refunding bonds. The Act of 1870 was itself motivated largely by the success of the Public Credit Act in clarifying redemption terms — which had established, by 1869, that 6% coin-payable coupons were now substantially above prevailing market rates and could be refinanced at the government's benefit.
